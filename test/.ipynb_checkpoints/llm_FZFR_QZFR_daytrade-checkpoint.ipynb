## 資料準備

```python
from smartmoneyconcepts import smc
from renko import Renko
import pandas as pd
from datetime import datetime, timedelta, time
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from collections import defaultdict
import math, re, itertools
from functools import reduce
from finmlkit.bar.base import TradesData
from finmlkit.feature.core.volume import VolumePro
from finmlkit.bar.kit import VolumeBarKit, TickBarKit, DollarBarKit, TimeBarKit, CUSUMBarKit
from finmlkit.feature.kit import Feature, FeatureKit, Compose
from finmlkit.feature import transforms as ts
from finmlkit.feature.base import SISOTransform, SIMOTransform, MISOTransform, BaseTransform, MIMOTransform
from finmlkit.feature.core.volatility import parkinson_range
from finmlkit.feature.core.momentum import roc
import talib

def prepare_data(type_of_data, data_name):
    result = type_of_data.split('/')[0]
    tmp = pd.read_csv(f'../index_data/{type_of_data}/{data_name}.csv')
    if result == 'shioaji':
        tmp['ts'] = pd.to_datetime(tmp['ts'])
        tmp = tmp.rename(columns=lambda x: x.lower())
    else:
        tmp['ts'] = pd.to_datetime(tmp['datetime'])
        tmp = tmp.rename(columns=lambda x: x.lower())
    
    return tmp

def aggregate_pv_list(pv_lists):
    if pv_lists.isna().all():
        return []
    
    combined_volume = defaultdict(float)
    for pv in pv_lists.dropna():
        if isinstance(pv, list):  # 確保是 list
            for item in pv:
                if isinstance(item, dict):
                    for price, vol in item.items():
                        combined_volume[float(price)] += float(vol)  # 轉 float 避免類型問題

    # 按價格降序排序並轉為 list of dict
    aggregated_pv = [{price: combined_volume[price]} for price in sorted(combined_volume, reverse=True)]
    return aggregated_pv

def create_cumulative_pv_data(pv_series):
    cumulative_data = {}
    all_previous_data = {}  # 存儲所有價位的累積量
    
    for timestamp, pv_list in pv_series.items():
        # 【新增清洗步驟】
        # 如果取出的 pv_list 不是列表 (例如是 NaN 的 float)，就強制視為空列表
        if not isinstance(pv_list, list):
            pv_list = []
        
        # 複製之前累積的數據
        current_cumulative = all_previous_data.copy()
        
        # 累加當前時間點的數據
        # (因為上面已經清洗過，這裡 pv_list 保證是 list，不會報錯)
        for item in pv_list:
            for price, volume in item.items():
                price_key = float(price)  # 確保價格是數值類型
                if price_key in current_cumulative:
                    current_cumulative[price_key] += volume
                else:
                    current_cumulative[price_key] = volume
        
        # 轉換為與原始pv_list相同的格式
        cumulative_list = [{price: volume} for price, volume in current_cumulative.items()]
        cumulative_data[timestamp] = cumulative_list
        
        # 更新累積數據用於下一個時間點
        all_previous_data = current_cumulative.copy()
    
    return pd.Series(cumulative_data, index=pv_series.index)

def calculate_liquidity_factors(bid_list, ask_list, close_list, vol_list):
    bids = np.array(bid_list)
    asks = np.array(ask_list)
    vols = np.array(vol_list)
    closes = np.array(close_list)
    valid_mask = (bids > 0) & (asks > 0) & (vols > 0)
    
    if not np.any(valid_mask):
        return 0
    
    # 3. 提取有效數據
    v_bids = bids[valid_mask]
    v_asks = asks[valid_mask]
    v_vols = vols[valid_mask]
    v_closes = closes[valid_mask]

    # A. 計算中間價
    mid_prices = (v_bids + v_asks) / 2

    # B. 計算單筆 Effective Spread (2 * |成交價 - 中間價|)
    eff_spreads = 2 * np.abs(v_closes - mid_prices)

    # 分子 1: 有效價差 * 成交量
    sum_weighted_eff = np.sum(eff_spreads * v_vols)

    return sum_weighted_eff

def calculate_price_factors(close_list, volume_list):
    """
    計算秒級價格因子。
    """
    try:
        sum_price_volume = sum((float(c)) * float(v) for c, v in zip(close_list, volume_list))
        sum_price_sq_volume = sum((float(c) ** 2) * float(v) for c, v in zip(close_list, volume_list))
        
        # 計算 pv_list：按 close 價格聚合成交量
        volume_by_close = defaultdict(float)
        for close, volume in zip(close_list, volume_list):
            volume_by_close[close] += float(volume)
        
        # 轉為 pv_list 格式並按 close 價格降序排序
        pv_list = [{close: volume} for close, volume in sorted(volume_by_close.items(), key=lambda x: x[0], reverse=True)]
        
        return pd.Series({
            'sum_price_volume': sum_price_volume,
            'sum_price_sq_volume': sum_price_sq_volume,
            'pv_list': pv_list
        })
    except (ValueError, ZeroDivisionError, TypeError, IndexError):
        return pd.Series({
            'sum_price_volume': np.nan,
            'sum_price_sq_volume': np.nan,
            'pv_list': np.nan
        })

def calculate_capital_factors(close_list, volume_list, tick_type_list):
    if not close_list or not volume_list or not tick_type_list:
        return pd.Series({
            'ticks_buy': 0, 'ticks_sell': 0,
            'volume_buy': 0.0, 'volume_sell': 0.0,
            'dollars_buy': 0.0, 'dollars_sell': 0.0,
            'volume_delta': 0.0,
            'monetary_delta': 0.0
        })

    # 2. 轉成 numpy（速度快 10~50 倍）
    c = np.array(close_list, dtype=float)
    v = np.array(volume_list, dtype=float)
    t = np.array(tick_type_list, dtype=int)

    # 3. 只保留有明確方向的 tick（1 或 -1），中性單不算買賣力
    active_mask = np.isin(t, [1, -1])
    if not np.any(active_mask):
        # 全是 0（中性單），所有買賣統計都是 0
        return pd.Series({
            'ticks_buy': 0, 'ticks_sell': 0,
            'volume_buy': 0.0, 'volume_sell': 0.0,
            'dollars_buy': 0.0, 'dollars_sell': 0.0,
            'volume_delta': 0.0,
            'monetary_delta': 0.0
        })

    c = c[active_mask]
    v = v[active_mask]
    t = t[active_mask]

    # 4. 建立買賣遮罩
    buy_mask  = (t == 1)
    sell_mask = (t == -1)

    # 5. 計算各項指標
    ticks_buy    = int(np.sum(buy_mask))
    ticks_sell   = int(np.sum(sell_mask))
    volume_buy   = float(np.sum(v[buy_mask]))
    volume_sell  = float(np.sum(v[sell_mask]))
    dollars_buy  = float(np.sum(c[buy_mask] * v[buy_mask]))
    dollars_sell = float(np.sum(c[sell_mask] * v[sell_mask]))

    volume_delta   = volume_buy - volume_sell
    monetary_delta = dollars_buy - dollars_sell

    return pd.Series({
        'ticks_buy'      : ticks_buy,
        'ticks_sell'     : ticks_sell,
        'volume_buy'     : volume_buy,
        'volume_sell'    : volume_sell,
        'dollars_buy'    : dollars_buy,
        'dollars_sell'   : dollars_sell,
        'volume_delta'   : volume_delta,
        'monetary_delta' : monetary_delta
    })
        
def calculate_volume_footprint(close_list, volume_list, tick_type_list):
    """
    计算每秒的价位成交量分布，按买/卖/中性区分。
    返回一个dict: {price: {'buy': vol, 'sell': vol, 'neutral': vol}}
    """
    if not close_list or not volume_list or not tick_type_list:
        return {}

    footprint = defaultdict(lambda: {'buy': 0.0, 'sell': 0.0, 'neutral': 0.0})
    
    for price, vol, tick in zip(close_list, volume_list, tick_type_list):
        if tick == 1:
            footprint[price]['buy'] += vol
        elif tick == -1:
            footprint[price]['sell'] += vol
        else:
            footprint[price]['neutral'] += vol
    
    # 转换为普通dict，便于后续聚合
    return dict(footprint)

def aggregate_footprints(footprints_series):
    if footprints_series.isna().all():
        return {}
    
    # 定義預設結構
    aggregated = defaultdict(lambda: {'buy': 0.0, 'sell': 0.0, 'neutral': 0.0})

    for footprint in footprints_series.dropna():
        if isinstance(footprint, dict):
            for price, vols in footprint.items():
                aggregated[price]['buy'] += vols.get('buy', 0.0)
                aggregated[price]['sell'] += vols.get('sell', 0.0)
                aggregated[price]['neutral'] += vols.get('neutral', 0.0)

    sorted_aggregated = {
        k: aggregated[k] 
        for k in sorted(aggregated.keys(), reverse=True) # 價格從高到低
    }
    
    return sorted_aggregated

def build_finmlkit_trade(df):
    origin_df = df.copy()
    origin_df.index = pd.to_datetime(origin_df['ts']) 
    
    return TradesData(
        ts=origin_df.index.view(np.int64),
        px=origin_df['close'].values,
        qty=origin_df['volume'].values,
        dt_index=origin_df.index,
        side=origin_df['tick_type'].values,
        timestamp_unit='ns' 
    )

def pre_process_by_finmlkit_volbar(df, volume_threshold):
    trade = build_finmlkit_trade(df)
    volume_bar_kit = VolumeBarKit(
        trades=trade,
        volume_ths=volume_threshold
    )

    return volume_bar_kit

def pre_process_by_finmlkit_tickbar(df, n):
    trade = build_finmlkit_trade(df)
    tick_bar_kit = TickBarKit(
        trades=trade,
        tick_count_thrs=n
    )

    return tick_bar_kit

def pre_process_by_finmlkit_dollarbar(df, dollar_threshold):
    trade = build_finmlkit_trade(df)
    dollar_bar_kit = DollarBarKit(
        trades=trade,
        dollar_thrs=dollar_threshold # ✅ 每 1,000,000 (1_000_000) 元成交金額形成一根 bar（可調整）
    )

    return dollar_bar_kit

def pre_process_by_finmlkit_timebar(df, time):
    trade = build_finmlkit_trade(df)
    time_bar_kit = TimeBarKit(
        trades=trade,
        period=pd.Timedelta(time) # ✅ 每 1 分鐘形成一根 bar，可改'1min', '5s', '10s', '1h' 等
    )
    
    return time_bar_kit

def pre_process_by_finmlkit_sumbar(df, window, min_sigma, sigma_mult):
    trade = build_finmlkit_trade(df)
    
    # 使用 log-returns 與 rolling std，window 可調（例如 200 ticks）
    prices = df['close'].astype(float)
    logret = np.log(prices).diff().fillna(0.0)
    
    # rolling std on returns (align with ticks)
    rolling_std = logret.rolling(window=window, min_periods=1, center=False).std().to_numpy(dtype=np.float64)
    
    # 如果 rolling_std 太小或為 0，會在 CUSUM 裡用 sigma_floor 修正
    sigma_vector = rolling_std 
    
    cusum_bar_kit = CUSUMBarKit(
        trades=trade,
        sigma=sigma_vector, # 1d np.ndarray 或常數
        sigma_floor=min_sigma,   # 最小 sigma（可視價格單位或 log-return 單位調）
        sigma_mult=sigma_mult      # lambda_th = sigma_mult * max(sigma, sigma_floor)
    )
    
    return cusum_bar_kit

def concate_finmlkit(list_of_obj_dict, name, **params):
    ohlcv_dict = {}
    direction_dict = {}
    size_dict = {}
    footprint = {}
    vp_dict = {}
    
    volpro = VolumePro(
        window_size=timedelta(minutes=params.get('min')),
        n_bins=params.get('nbin'),
        va_pct=params.get('va_pct')
    )

    for code, obj_list in list_of_obj_dict.items():
        ohlcv_list, direction_list, size_list, footprint_list, vp_list = [], [], [], [], []
        for item in obj_list:
            k = item.build_ohlcv()
            ohlcv_list.append(k)
            direction_list.append(item.build_directional_features())
            n_bars = len(item._close_indices) - 1
            theta_array = np.full(n_bars, 10, dtype=np.float64)
            size_list.append(item.build_trade_size_features(theta=theta_array))
            footprint_data = item.build_footprints()
            
            poc, hva, lva, pct_above = volpro.compute(k, footprint_data)
            df_vp_chunk = pd.DataFrame({
                'poc': poc,
                'hva': hva,
                'lva': lva,
                'pct_above': pct_above
            }, index=k.index)
            vp_list.append(df_vp_chunk)
            
            detail_footprint = footprint_data.get_df().reset_index(level=0, drop=True)

            footprint_list.append(detail_footprint)
        
        # 合併各項資料
        ohlcv_df = pd.concat(ohlcv_list) if ohlcv_list else pd.DataFrame()
        direction_df = pd.concat(direction_list) if direction_list else pd.DataFrame()
        size_df = pd.concat(size_list) if size_list else pd.DataFrame()
        footprint_df = pd.concat(footprint_list) if footprint_list else pd.DataFrame()
        vp_df = pd.concat(vp_list) if vp_list else pd.DataFrame()

        # 對時間索引整合
        if not ohlcv_df.empty:
            ohlcv_df = ohlcv_df.groupby(ohlcv_df.index).last().sort_index()
        if not direction_df.empty:
            direction_df = direction_df.groupby(direction_df.index).last().sort_index()
        if not size_df.empty:
            size_df = size_df.groupby(size_df.index).last().sort_index()
        if not footprint_df.empty:
            footprint_df = footprint_df.groupby(footprint_df.index).last().sort_index()
        if not vp_df.empty:
            vp_df = vp_df.groupby(vp_df.index).last().sort_index()

        # 存進字典
        suffix = f"_{name}_{code}"
        ohlcv_dict[code] = ohlcv_df.add_suffix(suffix)
        direction_dict[code] = direction_df.add_suffix(suffix)
        size_dict[code] = size_df.add_suffix(suffix)
        footprint[code] = footprint_df.add_suffix(suffix)
        vp_dict[code] = vp_df.add_suffix(suffix)

    # === 對齊多商品資料 === #
    def merge_all(dfs_dict):
        non_empty = [df for df in dfs_dict.values() if not df.empty]
        if not non_empty:
            return pd.DataFrame()
        merged = pd.concat(non_empty, axis=1)
        merged = merged.sort_index().ffill().dropna()
        return merged

    df_kbar = merge_all(ohlcv_dict)
    df_direction = merge_all(direction_dict)
    df_size = merge_all(size_dict)
    df_footprint = merge_all(footprint)
    df_vp = merge_all(vp_dict)

    return df_kbar, df_direction, df_size, df_footprint, df_vp

def pre_process_by_range(df, range_size=0.005):
    df = df.copy()
    df['ts'] = pd.to_datetime(df['ts'])
    df = df.sort_values('ts').reset_index(drop=True)
    df.rename(columns={'close': 'tick_close'}, inplace=True)
    
    # 計算買賣量
    df['buy_volume'] = np.where(df['tick_type'] == 1, df['volume'], 0)
    df['sell_volume'] = np.where(df['tick_type'] == 2, df['volume'], 0)

    # 初始化 bar_id
    prices = df['tick_close'].values
    n = len(df)
    bar_ids = np.zeros(n, dtype=int)
    current_bar_id = 0
    bar_open = prices[0]
    upper_bound = bar_open * (1 + range_size)
    lower_bound = bar_open * (1 - range_size)

    # 向量化生成 bar_id
    for i in range(1, n):
        price = prices[i]
        if price >= upper_bound or price <= lower_bound:
            current_bar_id += 1
            bar_open = price
            upper_bound = bar_open * (1 + range_size)
            lower_bound = bar_open * (1 - range_size)
        bar_ids[i] = current_bar_id

    df['bar_id'] = bar_ids

    # 按 bar_id 聚合，生成 K 線數據
    bar_stats = df.groupby('bar_id').agg(
        ts_range=('ts', 'first'),                    # K線開始時間
        open_range=('tick_close', 'first'),          # 開盤
        high_range=('tick_close', 'max'),            # 最高
        low_range=('tick_close', 'min'),             # 最低
        close_range=('tick_close', 'last'),          # 收盤
        volume_range=('volume', 'sum'),              # 總成交量
        buy_volume_range=('buy_volume', 'sum'),      # 主動買量
        sell_volume_range=('sell_volume', 'sum')     # 主動賣量
    ).reset_index()
    
    # 計算 order_imbalance
    total_vol = bar_stats['buy_volume_range'] + bar_stats['sell_volume_range']
    bar_stats['order_imbalance_range'] = np.where(
        total_vol > 0,
        (bar_stats['buy_volume_range'] - bar_stats['sell_volume_range']) / total_vol,
        0
    )

    # 設定 ts_range 為 index
    bar_stats = bar_stats.set_index('ts_range')

    # 最終欄位順序（保留 bar_id）
    final_columns = [
        'bar_id',
        'open_range', 'high_range', 'low_range', 'close_range',
        'volume_range', 'buy_volume_range', 'sell_volume_range',
        'order_imbalance_range'
    ]
    
    return bar_stats[final_columns]

def pre_process_by_renko(df, brick_size=0.5, renko_type='normal'):
    df = df.set_index('ts')
    r = Renko(df, brick_size=brick_size)
    rdf = r.renko_df(renko_type)
    rdf = rdf.rename(columns={
        'open': 'open_renko',
        'high': 'high_renko',
        'low': 'low_renko',
        'close': 'close_renko',
        'volume': 'volume_renko'
    })
    return rdf

def pre_process_by_second(df1, data_type, date_start, date_end):
    # 假设你的df已经加载到dataframe
    df1['ts'] = pd.to_datetime(df1['ts'])  # 将ts列转换为datetime类型

    # 按秒分组，使用 named aggregation
    df1 = df1.groupby(df1['ts'].dt.floor('s')).agg(
        close=('close', 'last'),
        volume=('volume', 'sum'), # 總成交量
        close_list=('close', list),
        volume_list=('volume', list), # 成交量列表
        bid_list=('bid_price', list),
        ask_list=('ask_price', list),
        bid_price=('bid_price', lambda x: tuple(sorted(filter(lambda price: price != 0, x)))),
        ask_price=('ask_price', lambda x: tuple(sorted(filter(lambda price: price != 0, x)))),
        tick_type=('tick_type', lambda x: [1 if t == 1 else -1 if t == 2 else 0 for t in x])
    ).reset_index()
    
    df1['eff_spread'] = df1.apply(lambda row: calculate_liquidity_factors(row['bid_list'], row['ask_list'], row['close_list'], row['volume_list']), axis=1)
    
    price_features = df1.apply(lambda row: calculate_price_factors(row['close_list'], row['volume_list']), axis=1)
    df1 = pd.concat([df1, price_features], axis=1)
    
    capital_features = df1.apply(lambda row: calculate_capital_factors(row['close_list'], row['volume_list'], row['tick_type']), axis=1)
    df1 = pd.concat([df1, capital_features], axis=1)
    
    # 新增：计算每秒的volume footprint
    df1['volume_footprint'] = df1.apply(lambda row: calculate_volume_footprint(row['close_list'], row['volume_list'], row['tick_type']), axis=1)

    df1.drop(columns=['tick_type', 'volume_list', 'bid_list', 'ask_list', 'close_list'], inplace=True)

    # 创建时间范围从開始到結束天數（或多个天数）
    time_range = pd.date_range(date_start, date_end, freq='s')

    # 将时间范围转换为DataFrame
    full_time_df = pd.DataFrame(time_range, columns=['ts'])
    
    if data_type == 's_day':
        # 通过检查时间是否在9:00:00到13:30:00之间来剔除跨天的数据
        valid_time_range = full_time_df['ts'].dt.time.between(pd.to_datetime('09:00:00').time(), pd.to_datetime('13:30:00').time())
        valid_time = full_time_df[valid_time_range]
        
    elif data_type == 'f_day':
        # 通过检查时间是否在9:00:00到13:30:00之间来剔除跨天的数据
        valid_time_range = full_time_df['ts'].dt.time.between(pd.to_datetime('08:45:00').time(), pd.to_datetime('13:45:00').time())
        valid_time = full_time_df[valid_time_range]
        
    elif data_type == 'f_night':
        t1 = pd.to_datetime('08:45:00').time()
        t2 = pd.to_datetime('13:45:00').time()
        t3 = pd.to_datetime('15:00:00').time()
        t4 = pd.to_datetime('05:00:00').time()
        
        # 提取時間部分並進行向量化比較
        time_series = full_time_df['ts'].dt.time
        valid_time = full_time_df[
            ((time_series >= t1) & (time_series <= t2)) |  # 日盤時間
            ((time_series >= t3) | (time_series <= t4))   # 夜盤時間
        ]

    # 合并df1和df2的结果，确保它们与mer_ori_data按秒对齐, 首先将df1和df2与mer_ori_data合并，使用'left'连接方式，以保留所有有效时间
    mer_ori_data = pd.merge(valid_time, df1, on='ts', how='left')

    # 设置'ts'为index
    mer_ori_data.set_index('ts', inplace=True)
    mer_ori_data = mer_ori_data.dropna()
    
    extra_df = mer_ori_data.resample('1min').agg({
        'eff_spread': 'sum',
        "ticks_buy": 'sum',
        "ticks_sell": 'sum',
        "volume_buy": 'sum',
        "volume_sell": 'sum',
        "dollars_buy": 'sum',
        "dollars_sell": 'sum',
        'sum_price_volume': 'sum',
        'sum_price_sq_volume': 'sum',
        'monetary_delta': 'sum',
        'volume_delta': 'sum',
        'pv_list': aggregate_pv_list,
        'volume_footprint': aggregate_footprints
    })
    
    # 為了配合合併分K, 需把時間+1分鐘
    extra_df.index = extra_df.index + pd.Timedelta(minutes=1)
    
    # 過濾掉 bid_price 或 ask_price 為空 tuple 的行 (漲停或跌停)
    mer_ori_data = mer_ori_data[(mer_ori_data['bid_price'].map(len) > 0) & (mer_ori_data['ask_price'].map(len) > 0)]
    
    return mer_ori_data.dropna(), extra_df.dropna()

def convert_ohlcv(df, freq=60):
    # 建立 session_type 與 session_start
    t1 = datetime.strptime("08:45", "%H:%M").time()
    t2 = datetime.strptime("13:45", "%H:%M").time()
    t3 = datetime.strptime("05:00", "%H:%M").time()
    t4 = datetime.strptime("15:00", "%H:%M").time()
    
    def classify_session(timestamps):
        try:
            # 1. 使用 .dt.tz 來檢查時區信息
            timestamps = timestamps.dt.tz_localize(None) if timestamps.dt.tz is not None else timestamps
            
            # 2. 使用 .dt.time 和 .dt.date 來提取時間和日期部分
            times = timestamps.dt.time
            dates = timestamps.dt.date

            # 初始化結果
            session_type = pd.Series("other", index=timestamps.index)
            session_start = pd.Series(pd.NaT, index=timestamps.index)
        
            # 日盤條件
            day_mask = (times >= t1) & (times <= t2)
            session_type.loc[day_mask] = "day"
            session_start.loc[day_mask] = pd.to_datetime(
                dates[day_mask].astype(str) + " " + t1.strftime("%H:%M:%S")
            )
    
            # 夜盤條件 (當天夜盤)
            night_mask1 = (times >= t4) & ~day_mask
            session_type.loc[night_mask1] = "night"
            session_start.loc[night_mask1] = pd.to_datetime(
                dates[night_mask1].astype(str) + " " + t4.strftime("%H:%M:%S")
            )
        
            # 夜盤條件 (前一天夜盤，時間 < t4)
            night_mask2 = (times < t4) & ~day_mask
            session_type.loc[night_mask2] = "night"
            prev_dates = (timestamps[night_mask2] - timedelta(days=1)).dt.date
            session_start.loc[night_mask2] = pd.to_datetime(
                prev_dates.astype(str) + " " + t4.strftime("%H:%M:%S")
            )

            return pd.DataFrame({"session_type": session_type, "session_start": session_start})

        except Exception as e:
            print(f"時間轉換出現錯誤: {e}")
            return pd.DataFrame({"session_type": None, "session_start": None}, index=timestamps.index)
    
    result = classify_session(df.index.to_series())
    df.loc[:, ["session_type", "session_start"]] = result[["session_type", "session_start"]]
    df = df[df["session_type"].isin(["day", "night"])]

    # 新增4: 補齊缺失的1分K資料
    def fill_missing_minutes(df_session, session_start, session_type):
        # 定義交易時段範圍
        if session_type == "day":
            start_time = datetime.combine(session_start.date(), t1)
            end_time = datetime.combine(session_start.date(), t2)
        else:  # night
            start_time = datetime.combine(session_start.date(), t4)
            end_time = datetime.combine(session_start.date() + timedelta(days=1), t3)

        # 生成完整的1分鐘時間序列
        full_time_index = pd.date_range(start=start_time, end=end_time, freq="1min")
        existing_times = df_session.index

        # 找出缺失的時間點
        missing_times = [t for t in full_time_index if t not in existing_times]
        
        if missing_times:
            # 為每個缺失時間點填充資料
            missing_data = []
            last_valid_row = None
            for t in missing_times:
                # 找到前一筆有效資料
                prev_time = t - timedelta(minutes=1)
                if prev_time in df_session.index:
                    last_valid_row = df_session.loc[prev_time]
                if last_valid_row is not None:
                    missing_data.append({
                        "ts": t,
                        "open": last_valid_row["close"],
                        "high": last_valid_row["close"],
                        "low": last_valid_row["close"],
                        "close": last_valid_row["close"],
                        "volume": 0,
                        "amount": 0,
                        "complete": True,
                        "session_type": session_type,
                        "session_start": session_start
                    })

            # 將缺失資料合併到原資料
            if missing_data:
                missing_df = pd.DataFrame(missing_data).set_index("ts")
                df_session = pd.concat([df_session, missing_df]).sort_index()

        return df_session

    # 按 session 分組並補齊缺失資料
    df_filled = []
    for session_start, session_data in df.groupby("session_start"):
        session_type = session_data["session_type"].iloc[0]
        session_data = fill_missing_minutes(session_data, session_start, session_type)
        df_filled.append(session_data)
        
    if df_filled:
        df = pd.concat(df_filled).sort_index()

    df.index = df.index - pd.Timedelta(minutes=1)

    # 設定 K 棒時間長度
    window = timedelta(minutes=freq)

    # 分段處理每個 session 的資料
    result = []

    for session_start, session_data in df.groupby("session_start"):
        current_time = session_start
        max_time = session_data.index.max()
        
        # 对 session 进行累积计算
        if not session_data.empty:
            session_data["acc_vol"] = session_data["volume"].cumsum()  # 在 session 内累积 volume
            session_data["acc_price_volume"] = session_data["sum_price_volume"].cumsum()  # 累积 sum_price_volume
            session_data["acc_price_sq_volume"] = session_data["sum_price_sq_volume"].cumsum()  # 累积 sum_price_sq_volume

        while current_time < max_time:
            next_time = current_time + window
            window_data = session_data[(session_data.index >= current_time) & (session_data.index < next_time)]

            if not window_data.empty:
                o = window_data["open"].iloc[0]
                h = window_data["high"].max()
                l = window_data["low"].min()
                c = window_data["close"].iloc[-1]
                v = window_data["volume"].sum()
                a = window_data["amount"].sum()
                complete = window_data.index[-1] >= next_time - timedelta(minutes=1)
                
                # 添加额外的列聚合
                e_spread = window_data["eff_spread"].sum()
                tb = window_data["ticks_buy"].sum()
                ts = window_data["ticks_sell"].sum()
                vb = window_data["volume_buy"].sum()
                vs = window_data["volume_sell"].sum()
                db = window_data["dollars_buy"].sum()
                ds = window_data["dollars_sell"].sum()
                sum_price_volume = window_data['sum_price_volume'].sum()
                sum_price_sq_volume = window_data['sum_price_sq_volume'].sum()
                monetary_delta_agg = window_data['monetary_delta'].sum()
                volume_delta_agg = window_data['volume_delta'].sum()
                pv_list_agg = aggregate_pv_list(window_data['pv_list'])
                vol_footprint_agg = aggregate_footprints(window_data['volume_footprint'])
                acc_vol = window_data["acc_vol"].sum()
                acc_price_vol = window_data["acc_price_volume"].sum()
                acc_price_sq_volume = window_data["acc_price_sq_volume"].sum()

                result.append({
                    "ts": current_time,
                    "open": o,
                    "high": h,
                    "low": l,
                    "close": c,
                    "volume": v,
                    "amount": a,
                    'eff_spread': e_spread,
                    "ticks_buy": tb,
                    "ticks_sell": ts,
                    "volume_buy": vb,
                    "volume_sell": vs,
                    "dollars_buy": db,
                    "dollars_sell": ds,
                    "sum_price_volume": sum_price_volume,
                    "sum_price_sq_volume": sum_price_sq_volume,
                    "monetary_delta": monetary_delta_agg,
                    "volume_delta": volume_delta_agg,
                    "pv_list": pv_list_agg,
                    "volume_footprint": vol_footprint_agg,
                    "acc_volume": acc_vol,
                    "acc_price_volume": acc_price_vol,
                    "acc_price_sq_volume": acc_price_sq_volume,
                    "complete": complete
                })

            current_time = next_time
            
    # 建立新的 DataFrame
    agg_df = pd.DataFrame(result)
    agg_df.set_index("ts", inplace=True, drop=False)
    agg_df = agg_df.shift(1).dropna()

    return agg_df

def combine_daily_k_bars(df):
    df = df.copy()
    df.index = pd.to_datetime(df.index)
    df['date'] = df.index.date
    df['time'] = df.index.time
    
    combined_data = []
    
    for date, group in df.groupby('date'):
        aggregated_row = {
            'date': date,
            'open': group['open'].iloc[0],
            'high': group['high'].max(),
            'low': group['low'].min(),
            'close': group['close'].iloc[-1],
            'volume': group['volume'].sum(),
            'amount': group['amount'].sum(),
            'complete': True,
            'eff_spread': group["eff_spread"].sum(),
            "ticks_buy": group["ticks_buy"].sum(),
            "ticks_sell": group["ticks_sell"].sum(),
            "volume_buy": group["volume_buy"].sum(),
            "volume_sell": group["volume_sell"].sum(),
            "dollars_buy": group["dollars_buy"].sum(),
            "dollars_sell": group["dollars_sell"].sum(),
            'monetary_delta': group['monetary_delta'].sum(),
            'volume_delta': group['volume_delta'].sum(),
            'acc_price_volume': group['acc_price_volume'].iloc[-1],
            'acc_price_sq_volume': group['acc_price_sq_volume'].iloc[-1],
            'acc_volume': group['acc_volume'].iloc[-1],
            'pv_list':  aggregate_pv_list(group['pv_list']),
            'volume_footprint': aggregate_footprints(group['volume_footprint']),
        }
        
        combined_data.append(aggregated_row)
    
    # 创建结果DataFrame
    combined = pd.DataFrame(combined_data)
    combined['ts'] = pd.to_datetime(combined['date'])
    combined = combined.set_index('ts')
    combined = combined.drop('date', axis=1)
    
    return combined.shift(1).dropna()

def process_multiple_datasets_by_second(dataset1, dataset2, codes):
    dfs_by_code = {code: [] for code in codes}
    dfk_by_code = {code: [] for code in codes}
    extra_by_code = {code: [] for code in codes}
    
    # 處理 tick 資料
    for type_of_data, data_name, date_start, date_end, data_type, *params in dataset1:
        origin_df = prepare_data(type_of_data, data_name)
        df, extra_data = pre_process_by_second(origin_df, data_type, date_start, date_end)
        
        for code in codes:
            if data_name.startswith(code):
                dfs_by_code[code].append(df)
                extra_by_code[code].append(extra_data)
                break

    # 處理 1分K 資料
    for type_of_data, data_name in dataset2:
        df = prepare_data(type_of_data, data_name).set_index('ts')
        
        for code in codes:
            if data_name.startswith(code + 'k'):
                dfk_by_code[code].append(df)
                break

    # 合併各商品的 dataframe
    final_dfs = {}
    final_dfks = {}

    for code in codes:
        # 合併同一商品的多個資料段
        df = pd.concat(dfs_by_code[code]) if dfs_by_code[code] else pd.DataFrame()
        dfk = pd.concat(dfk_by_code[code]) if dfk_by_code[code] else pd.DataFrame()
        extra = pd.concat(extra_by_code[code]) if extra_by_code[code] else pd.DataFrame()

        # groupby 處理重複時間
        if not df.empty:
            df = df.groupby(df.index).last().sort_index().dropna()
        if not dfk.empty:
            dfk = dfk.groupby(dfk.index).last().sort_index().dropna()

        # 1k 資料加上 extra 資料（若存在）
        if not dfk.empty and not extra.empty:
            dfk = pd.merge(dfk, extra, left_index=True, right_index=True, how="inner")

        final_dfs[code] = df
        final_dfks[code] = dfk

    # === 合併所有商品成一張 df（依 index inner join）===
    merged_df = None
    for code in codes:
        if final_dfs[code].empty:
            continue
        if merged_df is None:
            merged_df = final_dfs[code].copy()
            merged_df = merged_df.add_suffix(f'_{code}')
        else:
            df_tmp = final_dfs[code].add_suffix(f'_{code}')
            merged_df = pd.merge(merged_df, df_tmp, left_index=True, right_index=True, how='inner')

    return merged_df, final_dfks

def process_multiple_datasets_by_custom(dataset1, codes):
    # 建立每個 bar 類型的 dict
    bar_types = ["tick", "vol", "time", "sum", "dollar", "range", "renko"]
    bar_dict = {btype: {code: [] for code in codes} for btype in bar_types}

    # 處理 tick 資料
    for type_of_data, data_name, _, _, _, tick_bar, volume_bar, time_bar, sum_bar, dollar_bar, range_bar, renko_bar in dataset1:
        origin_df = prepare_data(type_of_data, data_name)
        
        if 'tick_type' not in origin_df.columns:
            # 指數填 0 (無買賣方向)
            origin_df['tick_type'] = np.zeros(len(origin_df), dtype=np.int8)
        else:
            # 確保 tick_type 是整數 (有時可能是 float)
            origin_df['tick_type'] = origin_df['tick_type'].fillna(0).astype(np.int8)
            
        # 檢查並補齊 volume
        if 'volume' not in origin_df.columns:
            # 指數填 0
            origin_df['volume'] = np.zeros(len(origin_df), dtype=np.float64)
        else:
            origin_df['volume'] = origin_df['volume'].fillna(0)
        
        tick_data   = pre_process_by_finmlkit_tickbar(origin_df, tick_bar.get('n', 50))
        volume_data = pre_process_by_finmlkit_volbar(origin_df, volume_bar.get('volume_threshold', 1000))
        time_data   = pre_process_by_finmlkit_timebar(origin_df, time_bar.get('time', '1min'))
        sum_data    = pre_process_by_finmlkit_sumbar(origin_df, sum_bar.get('window', 200), sum_bar.get('min_sigma', 5e-4), sum_bar.get('sigma_mult', 2))
        dollar_data = pre_process_by_finmlkit_dollarbar(origin_df, dollar_bar.get('dollar_threshold', 1_000_000))
        range_data  = pre_process_by_range(origin_df, range_bar.get('range_size', 0.005))
        renko_data  = pre_process_by_renko(origin_df, renko_bar.get('brick_size', 0.5), renko_bar.get('type', 'normal'))
        
        for code in codes:
            if data_name.startswith(code):
                bar_dict["tick"][code].append(tick_data)
                bar_dict["vol"][code].append(volume_data)
                bar_dict["time"][code].append(time_data)
                bar_dict["sum"][code].append(sum_data)
                bar_dict["dollar"][code].append(dollar_data)
                bar_dict["range"][code].append(range_data)
                bar_dict["renko"][code].append(renko_data)
                break

    # === 這裡不 concat，只是簡單複製 === #
    final_bars = bar_dict

    # === 合併 range / renko 成 DataFrame === #
    def merge_multi_bars(bar_data_dict):
        non_empty = [(code, df) for code, df in bar_data_dict.items() if not isinstance(df, list) and not df.empty]
        if not non_empty:
            return pd.DataFrame()
        merged = reduce(
            lambda left, right: pd.merge(
                left[1].add_suffix(f"_{left[0]}"),
                right[1].add_suffix(f"_{right[0]}"),
                left_index=True,
                right_index=True,
                how='outer'
            ),
            non_empty
        )
        merged = merged.sort_index().ffill().dropna()
        return merged

    # 因為 range/renko 的 bar 是 DataFrame（不是 finmlkit 物件）
    df_rangebar = merge_multi_bars(final_bars["range"])
    df_renkobar = merge_multi_bars(final_bars["renko"])

    return df_rangebar, df_renkobar, final_bars

CODE = ['FZFR1', 'QZFR1']
INDEX = []
session = 'f_day'
time_bar = '1440'
params1 = {
    'tick_bar': {'n': 120},
    'volume_bar': {'volume_threshold': 150},
    'time_bar': {'time': f'{time_bar}min'},
    'sum_bar': {'window': 200, 'min_sigma': 7.656013e-05, 'sigma_mult': 3.791},
    'dollar_bar': {'dollar_threshold': 100000},
    'range_bar': {'range_size': 0.007},
    'renko_bar': {'brick_size': 1}
}

params2 = {
    'tick_bar': {'n': 5},
    'volume_bar': {'volume_threshold': 10},
    'time_bar': {'time': f'{time_bar}min'},
    'sum_bar': {'window': 200, 'min_sigma': 4.003475e-04, 'sigma_mult': 4.061},
    'dollar_bar': {'dollar_threshold': 35000},
    'range_bar': {'range_size': 0.005},
    'renko_bar': {'brick_size': 2.5}
}

params3 = {
    'tick_bar': {'n': 5},
    'volume_bar': {'volume_threshold': 10},
    'time_bar': {'time': f'{time_bar}min'},
    'sum_bar': {'window': 200, 'min_sigma': 4.003475e-04, 'sigma_mult': 4.061},
    'dollar_bar': {'dollar_threshold': 35000},
    'range_bar': {'range_size': 0.005},
    'renko_bar': {'brick_size': 2.5}
}

params4 = {
    'tick_bar': {'n': 5},
    'volume_bar': {'volume_threshold': 10},
    'time_bar': {'time': f'{time_bar}min'},
    'sum_bar': {'window': 200, 'min_sigma': 4.003475e-04, 'sigma_mult': 4.061},
    'dollar_bar': {'dollar_threshold': 35000},
    'range_bar': {'range_size': 0.005},
    'renko_bar': {'brick_size': 2.5}
}

dataset1 = [
    ('shioaji/2025_0912', f'{CODE[0]}', '2025-09-01 08:45:00', '2025-12-17 13:45:00', f'{session}', *params1.values()),
    ('shioaji/2025_0912', f'{CODE[1]}', '2025-09-01 08:45:00', '2025-12-17 13:45:00', f'{session}', *params1.values()),
]

dataset2 = [
    ('shioaji/2025_0912', f'{CODE[0]}k'),
    ('shioaji/2025_0912', f'{CODE[1]}k'),
]

dataset3 = [
    # ('shioaji/2024_0111', f'{INDEX[0]}', '2024-01-02 9:00:00', '2025-11-10 13:30:00', f'{session}', *params4.values()),
]

dataset4 = [
    # ('shioaji/2024_0111', f'{INDEX[0]}k'),
]

args = {
    'min': 1440 * 5,
    'nbin': None,
    'va_pct': 68.34
}

# 放入 (貴的, 便宜的), 數值大的在ratio分母, 使得數值可以相除在0~1
df, dfk = process_multiple_datasets_by_second(dataset1, dataset2, CODE)

## Code
# df_range, df_renko, final_bars = process_multiple_datasets_by_custom(dataset1, CODE)
# df_time_k, df_time_direct, df_time_size, df_time_footprint, df_vp = concate_finmlkit(final_bars["time"], 'time', **args)
# df_tick_k, df_tick_direct, df_tick_size, df_tick_footprint = concate_finmlkit(final_bars["tick"], 'tick', **args)
# df_vol_k, df_vol_direct, df_vol_size, df_vol_footprint = concate_finmlkit(final_bars["vol"], 'volume')
# df_sum_k, df_sum_direct, df_sum_size, df_sum_footprint = concate_finmlkit(final_bars["sum"], 'sum')
# df_dollar_k, df_dollar_direct, df_dollar_size, df_dollar_footprint = concate_finmlkit(final_bars["dollar"], 'dollar')

## Index
# df_index_range, df_index_renko, index_final_bars = process_multiple_datasets_by_custom(dataset3, INDEX)
# df_index_time_k, df_index_time_direct, df_index_time_size, df_index_time_footprint, df_index_vp = concate_finmlkit(index_final_bars["time"], 'time', **args)
```

```python
def engineer_features(df, code_suffix):
    df = df.copy()
    cols = [col for col in df.columns if col.endswith(code_suffix)]
    df_stock = df[cols]
    
    result_df = pd.DataFrame(index=df.index)
    date_grouper = df.index.date

    tb = df_stock[f'ticks_buy{code_suffix}']
    ts = df_stock[f'ticks_sell{code_suffix}']
    vb = df_stock[f'volume_buy{code_suffix}']
    vs = df_stock[f'volume_sell{code_suffix}']
    db = df_stock[f'dollars_buy{code_suffix}']
    ds = df_stock[f'dollars_sell{code_suffix}']
    eff_spread = df_stock[f'eff_spread{code_suffix}']
    vd = df_stock[f'volume_delta{code_suffix}']   
    mvd = df_stock[f'monetary_delta{code_suffix}']
    acc_pv = df_stock[f'acc_price_volume{code_suffix}']
    acc_vol = df_stock[f'acc_volume{code_suffix}']
    acc_p2v = df_stock[f'acc_price_sq_volume{code_suffix}']
    pv = df_stock[f'pv_list{code_suffix}']

    # 提取原始數據
    result_df[f'vd{code_suffix}'] = vd
    result_df[f'mvd{code_suffix}'] = mvd
    result_df[f'net_sell_ratio{code_suffix}'] = abs(vd) / (vb + vs)
    
    # CVD (累積成交量 Delta): 每天從 0 開始累積
    result_df[f'cvd{code_suffix}'] = vd.groupby(date_grouper).cumsum()
    result_df[f'cmvd{code_suffix}'] = mvd.groupby(date_grouper).pct_change(fill_method=None)

    day_acc_pv = acc_pv.groupby(date_grouper).cumsum()
    day_acc_vol = acc_vol.groupby(date_grouper).cumsum()
    day_acc_p2v = acc_p2v.groupby(date_grouper).cumsum()
    day_eff_spread = eff_spread.groupby(date_grouper).cumsum()

    # 按日期分组计算cumulative_pv
    cumulative_results = []
    for date, group_indices in df.groupby(date_grouper).indices.items():
        daily_result = create_cumulative_pv_data(pv.iloc[group_indices])  
        
        # 将结果分配给对应的索引
        cumulative_results.extend(daily_result)

    result_df[f'cumulative_pv{code_suffix}'] = cumulative_results

    # 計算 VWAP
    vwap = day_acc_pv / day_acc_vol.replace(0, np.nan)
    
    # 計算 VWAP 標準差 (Standard Deviation) 公式: E[x^2] - (E[x])^2
    vwap_variance = (day_acc_p2v / day_acc_vol.replace(0, np.nan)) - (vwap ** 2)
    vwap_std = np.sqrt(vwap_variance.clip(lower=0)) # clip 避免浮點數誤差導致負數
    
    result_df[f'vwap{code_suffix}'] = vwap
    result_df[f'vwap_std{code_suffix}'] = vwap_std
    
    ti = (tb - ts) / (tb + ts)
    result_df[f'tick_imbalance{code_suffix}'] = ti
    
    di = (db - ds) / (db + ds)
    result_df[f'dollar_imbalance{code_suffix}'] = di
    
    vi = (vb - vs) / (vb + vs)
    result_df[f'volume_imbalance{code_suffix}'] = vi

    buy_tick_ratio   = tb / (tb + ts)
    buy_dollar_ratio = db / (db + ds)
    buy_volume_ratio = vb / (vb + vs)
    
    # (1) 金額版
    sm_bias_dollars = buy_dollar_ratio - buy_tick_ratio
    result_df[f'bias_dollars{code_suffix}'] = sm_bias_dollars.fillna(0)

    # (2) 次數版（反映買方是否更積極）
    sm_bias_ticks = buy_tick_ratio - 0.5
    result_df[f'bias_ticks{code_suffix}'] = sm_bias_ticks.fillna(0)

    # (3) 量能版（大戶通常單筆量大）
    sm_bias_volume = buy_volume_ratio - buy_tick_ratio
    result_df[f'bias_volume{code_suffix}'] = sm_bias_volume.fillna(0)
    
    avg_ticket_size = (db + ds) / (tb + ts)
    result_df[f'avg_ticket_size{code_suffix}'] = avg_ticket_size.fillna(0)

    avg_volume_per_trade = (vb + vs) / (tb + ts)
    result_df[f'avg_volume_per_trade{code_suffix}'] = avg_volume_per_trade.fillna(0)

    liquidity_cum = (vb + vs).replace(0, np.nan) / day_eff_spread
    liquidity_cum = liquidity_cum.replace([np.inf, -np.inf], 0).fillna(0)
    result_df[f'liqudity{code_suffix}'] = np.sqrt(liquidity_cum)
    
    return result_df.dropna()

engineered_dfs = []
for code in CODE:
    suffix = f'_{code}'
    engineered = engineer_features(df_timek1, code_suffix=suffix)
    engineered_dfs.append(engineered)

feature = pd.concat(engineered_dfs, axis=1).sort_index()
```

## LLM

```python
import time
from langchain_openai import ChatOpenAI
from langchain.memory import ConversationBufferWindowMemory
from langgraph.graph import StateGraph, END
from typing import TypedDict, Dict, List, Literal, Optional
import json, os, re, textwrap

########################################
## 定義 Graph 的 State
########################################

class GraphState(TypedDict):
    start_time: str  # 窗口開始時間
    end_time: str  # 窗口結束時間
    current_volatility: str
    volatility_result: str
    volatility_memory: Optional[ConversationBufferWindowMemory]
    current_microstructure: str
    microstructure_result: str
    microstructure_memory: Optional[ConversationBufferWindowMemory]
    current_pv: str
    pv_result: str
    pv_memory: Optional[ConversationBufferWindowMemory]
    current_footprint: str
    footprint_result: str
    footprint_memory: Optional[ConversationBufferWindowMemory] 
    analysis_result: str
    entry_memory: Optional[ConversationBufferWindowMemory] 
    llm: Optional[ChatOpenAI]  # 共享的 LLM 實例
    system_prompts: Dict[str, str]  # 系統提示詞字典

########################################
## 輔助函數
########################################

def format_chat_history(chat_history):
    """格式化聊天歷史為字串"""
    if not chat_history:
        return "（無歷史對話）"
    
    formatted_history = ""
    for msg in chat_history:
        # 判斷訊息類型
        if hasattr(msg, '__class__'):
            msg_type = msg.__class__.__name__
        else:
            msg_type = type(msg).__name__
        
        # 根據類型添加前綴
        if msg_type == "HumanMessage" or msg_type == "AIMessage":
            prefix = "分析期: " if msg_type == "HumanMessage" else "分析結果: "
            formatted_history += f"{prefix}{msg.content}\n"
        elif isinstance(msg, dict):
            role = msg.get("role", "unknown")
            prefix = "分析期: " if role in ["user", "human"] else "分析結果: "
            formatted_history += f"{prefix}{msg.get('content', '')}\n"
        else:
            formatted_history += f"{str(msg)}\n"
    
    return formatted_history

def clean_think_content(text: str) -> str:
    """
    清理包含 think 標籤的文字內容
    處理完整的 <think>...</think> 或只有 </think> 結尾的情況
    """
    # 同時處理兩種模式：
    # 1. <think>...任意內容...</think>
    # 2. 任意內容...</think>（從開頭到 </think>）
    pattern = r"(?:<think>)?.*?</think>"
    cleaned_text = re.sub(pattern, "", text, flags=re.DOTALL)
    
    return cleaned_text.strip()

def create_cumulative_pv_data(pv_series):
    cumulative_data = {}
    all_previous_data = {}  # 存儲所有價位的累積量
    
    for timestamp, pv_list in pv_series.items():
        # 【新增清洗步驟】
        # 如果取出的 pv_list 不是列表 (例如是 NaN 的 float)，就強制視為空列表
        if not isinstance(pv_list, list):
            pv_list = []
        
        # 複製之前累積的數據
        current_cumulative = all_previous_data.copy()
        
        # 累加當前時間點的數據
        # (因為上面已經清洗過，這裡 pv_list 保證是 list，不會報錯)
        for item in pv_list:
            for price, volume in item.items():
                price_key = float(price)  # 確保價格是數值類型
                if price_key in current_cumulative:
                    current_cumulative[price_key] += volume
                else:
                    current_cumulative[price_key] = volume
        
        # 轉換為與原始pv_list相同的格式
        cumulative_list = [{price: volume} for price, volume in current_cumulative.items()]
        cumulative_data[timestamp] = cumulative_list
        
        # 更新累積數據用於下一個時間點
        all_previous_data = current_cumulative.copy()
    
    return pd.Series(cumulative_data, index=pv_series.index)

def format_pv_data(pv, close):
    data_list = []
    for item in pv:
        for p, v in item.items():
            data_list.append((float(p), float(v)))
    
    # 確保按價格由低到高排序
    data_list.sort(key=lambda x: x[0], reverse=True)

    # 2. 找出最大成交量 (作為繪圖的 100% 基準)
    if not data_list:
        return "無數據"
    
    max_vol = max(item[1] for item in data_list)
    max_bar_width = 40  # 設定長條圖最大長度 (字元數)

    # 3. 建立結果字串
    result_lines = []
    result_lines.append("=" * 70)
    result_lines.append(f"籌碼峰 (當前最新收盤價: {close})")
    result_lines.append(f"{'價格':<8} | {'籌碼分佈':<42} | {'量':<6}")
    result_lines.append("-" * 70)

    # 4. 逐行繪製
    for price, volume in data_list:
        # 計算長條圖長度
        bar_len = int((volume / max_vol) * max_bar_width)
        bar_str = "█" * bar_len
        
        # 標記特殊狀態
        marker = ""
        row_style = ""
        
        # 判斷是否為當前價格 (使用小數點誤差容許比較)
        is_current = abs(price - close) < 0.001
        
        # 判斷是否為籌碼峰 (例如大於最大量的 80%)
        is_peak = volume > (max_vol * 0.8)
        
        # 判斷是否為真空區 (例如小於最大量的 25%)
        is_lva = volume < (max_vol * 0.25)

        # 組合顯示字串
        vol_str = f"{int(volume)}"
        
        if is_current:
            # 當前價位：加上箭頭強調
            marker = f" <--- Close @ {price}"
            # 為了讓當前行更顯眼，可以改變 bar 的視覺 (如果環境支援) 或加強標示
            result_lines.append(f"{price:>7.2f} | {bar_str:<42} | {vol_str:<6} {marker}")
        else:
            # 根據 LVA 或 Peak 加入簡單註解 (選用)
            note = ""
            if is_peak: note = "(峰)"
            if is_lva:  note = "." # 用小點標示相對弱勢區
            
            result_lines.append(f"{price:>7.2f} | {bar_str:<42} | {vol_str:<6} {note}")

    result_lines.append("=" * 70)
    
    # 回傳單一字串以便列印
    return "\n".join(result_lines)

def format_footprint(footprint, close):    
    result_dict = {
        ts.strftime('%H:%M'): value 
        for ts, value in footprint.items()
    }
    sorted_times = sorted(result_dict.keys())
    
    # 建立結果字串
    result_lines = []
    result_lines.append("=" * 60)
    result_lines.append("成交量足跡分析")
    result_lines.append("=" * 60)
    
    for idx, (time_key, current_close) in enumerate(zip(sorted_times, close), 1):
        price_dict = result_dict[time_key]
        
        # 標題行
        result_lines.append(f"\n[{idx}] 時間: {time_key}")
        result_lines.append("-" * 60)
        
        # 計算該時間段的統計
        total_buy = sum(v.get('buy', 0) for v in price_dict.values())
        total_sell = sum(v.get('sell', 0) for v in price_dict.values())
        total_neutral = sum(v.get('neutral', 0) for v in price_dict.values())
        total_volume = total_buy + total_sell + total_neutral
        
        # -------------------------------------------------------
        # 【修改點】在統計數據區塊中加入收盤價
        # -------------------------------------------------------
        result_lines.append(f"總項目數: {len(price_dict)} | 總成交量: {total_volume:.1f} | 收盤價: {current_close}")
        result_lines.append(f"買量: {total_buy:.1f} | 賣量: {total_sell:.1f} | 中立量: {total_neutral:.1f}")
        result_lines.append("-" * 60)
        
        # 按價格降序顯示詳細數據
        sorted_prices = sorted(price_dict.items(), key=lambda x: float(x[0]), reverse=True)
        
        for i, (price, volume_data) in enumerate(sorted_prices, 1):
            buy = volume_data.get('buy', 0)
            sell = volume_data.get('sell', 0)
            neutral = volume_data.get('neutral', 0)
            
            # 決定顯示符號
            net = buy - sell
            if net > 0:
                symbol = "↑"  # 買盤較強
            elif net < 0:
                symbol = "↓"  # 賣盤較強
            else:
                symbol = "→"  # 平衡
            
            result_lines.append(f"  {i:3d}. 價格: {price:>8} {symbol} "
                              f"買:{buy:>8.1f} 賣:{sell:>8.1f} 中立:{neutral:>8.1f}")
    
    # 將所有行合併為一個字串並回傳
    return "\n".join(result_lines)

########################################
## 建立 Memory
########################################

def create_memory(N=3):
    return ConversationBufferWindowMemory(
        return_messages=True,
        k=N
    )

########################################
## 初始化函數 - 創建初始狀態
########################################

def initialize_state(model, url, N, system_prompts) -> GraphState:
    return {
        "start_time": "",
        "end_time": "",
        "current_volatility": "",
        "volatility_result": "",
        "volatility_memory": create_memory(N=N),
        "current_microstructure": "",
        "microstructure_result": "",
        "microstructure_memory": create_memory(N=N),
        "current_pv": "",
        "pv_result": "",
        "pv_memory": create_memory(N=N),
        "current_footprint": "",
        "footprint_result": "",
        "footprint_memory": create_memory(N=N),
        "analysis_result": "",
        "llm": ChatOpenAI(base_url=url, api_key="sk-no-key-required", model=model, temperature=0),
        "entry_memory": create_memory(N=N),
        "system_prompts": system_prompts
    }

########################################
## 定義節點函數
########################################

def check_pv(state: GraphState):
    llm = state["llm"]
    current_prompt = state["system_prompts"]["prompt_check_pv"]
    
    # 从 memory 中加载历史
    memory = state["pv_memory"]
    history_dict = memory.load_memory_variables({})
    chat_history = history_dict.get('history', [])
    formatted_history = format_chat_history(chat_history)

    pv_prompt = textwrap.dedent(f"""
        請根據持當前輸入時刻的籌碼峰資料, 進行籌碼峰分析:
        **分析期**: {state["start_time"]} to {state["end_time"]}
        
        **當前籌碼峰**:
        {state["current_pv"]}
        
        **歷史分析**:
        {formatted_history}
        
        請使用以上籌碼峰資料, 依照規定進行分析與輸出。
    """).strip()
    
    reply = llm.invoke([
        {"role": "system", "content": current_prompt},
        {"role": "user", "content": pv_prompt}
    ])
    
    state["pv_result"] = clean_think_content(reply.content)
    return state

def check_footprint(state: GraphState):
    llm = state["llm"]
    current_prompt = state["system_prompts"]["prompt_check_footprint"]
    
    # 从 memory 中加载历史
    memory = state["footprint_memory"]
    history_dict = memory.load_memory_variables({})
    chat_history = history_dict.get('history', [])
    formatted_history = format_chat_history(chat_history)
    
    footprint_prompt = textwrap.dedent(f"""
        請根據持當前輸入時刻的成交量足跡資料, 進行成交量足跡分析:
        **分析期**: {state["start_time"]} to {state["end_time"]}
        
        **當前成交量足跡**:
        {state["current_footprint"]}
        
        **歷史分析**:
        {formatted_history}
        
        請使用以上成交量足跡資料, 依照規定進行分析與輸出。
    """).strip()
    
    reply = llm.invoke([
        {"role": "system", "content": current_prompt},
        {"role": "user", "content": footprint_prompt}
    ])

    state['footprint_result'] = clean_think_content(reply.content)
    return state

def check_microstructure(state: GraphState):
    llm = state["llm"]
    current_prompt = state["system_prompts"]["prompt_check_microstructure"]
    
    # 从 memory 中加载历史
    memory = state["microstructure_memory"]
    history_dict = memory.load_memory_variables({})
    chat_history = history_dict.get('history', [])
    formatted_history = format_chat_history(chat_history)
    
    microstructure_prompt = textwrap.dedent(f"""
        請根據當前輸入的市場盤口資料, 進行數值分析:
        **分析期**: {state["start_time"]} to {state["end_time"]}

        **市場盤口資料**:
        {state["current_microstructure"]}
        
        **歷史分析**:
        {formatted_history}

        請請使用市場盤口資料, 依照規定進行分析與結果輸出。
    """).strip()
    
    reply = llm.invoke([
        {"role": "system", "content": current_prompt},
        {"role": "user", "content": microstructure_prompt}
    ])
    
    state['microstructure_result'] = clean_think_content(reply.content)
    return state

def check_volatility(state: GraphState):
    llm = state["llm"]
    current_prompt = state["system_prompts"]["prompt_check_volatility"]
    
    # 从 memory 中加载历史
    memory = state["volatility_memory"]
    history_dict = memory.load_memory_variables({})
    chat_history = history_dict.get('history', [])
    formatted_history = format_chat_history(chat_history)
    
    volatility_prompt = textwrap.dedent(f"""
        請根據當前輸入的波動率資料, 進行數值分析:
        **分析期**: {state["start_time"]} to {state["end_time"]}

        **相關資料**:
        {state["current_volatility"]}
        
        **歷史分析**:
        {formatted_history}

        請請使用相關資料, 依照規定進行分析與結果輸出。
    """).strip()
    
    reply = llm.invoke([
        {"role": "system", "content": current_prompt},
        {"role": "user", "content": volatility_prompt}
    ])
    
    state['volatility_result'] = clean_think_content(reply.content)
    return state

def check_entry(state: GraphState):
    llm = state["llm"]
    current_prompt = state["system_prompts"]["prompt_check_entry"]
    footprint = state["footprint_result"]
    pv = state["pv_result"]
    volatility = state["volatility_result"]
    microstructure = state["microstructure_result"]
    
    # 从 memory 中加载历史
    memory = state["entry_memory"]
    history_dict = memory.load_memory_variables({})
    chat_history = history_dict.get('history', [])
    formatted_history = format_chat_history(chat_history)

    position_prompt = textwrap.dedent(f"""
        請根據傳入的資料進行數據分析:
        **分析期**: {state["start_time"]} to {state["end_time"]}

        **波動性**
        {volatility}
        
        **盤口微觀結構**:
        {microstructure}
        
        **籌碼峰**:
        {pv}
        
        **成交量足跡圖**:
        {footprint}
        
        **歷史分析**:
        {formatted_history}
        
        請分析當前要進行甚麼操作, 並依照規定進行輸出。
    """).strip()
    
    reply = llm.invoke([
        {"role": "system", "content": current_prompt},
        {"role": "user", "content": position_prompt}
    ])

    state["analysis_result"] = clean_think_content(reply.content)
    return state

########################################
## 建立 Graph
########################################

def build_graph():
    graph = StateGraph(GraphState)
    graph.add_node("check_microstructure", check_microstructure)
    graph.add_node("check_volatility", check_volatility)
    graph.add_node("check_pv", check_pv)
    graph.add_node("check_footprint", check_footprint)
    graph.add_node("check_entry", check_entry)
    
    graph.set_entry_point("check_volatility")
    graph.add_edge("check_volatility", "check_microstructure")
    graph.add_edge("check_microstructure", "check_pv")
    graph.add_edge("check_pv", "check_footprint")
    graph.add_edge("check_footprint", "check_entry")
    graph.add_edge("check_entry", END)

    return graph.compile()

########################################
## 主流程
########################################

def main(df_all, features, code, json_file, model, url, system_prompts, k_lookback=10, N=5, **params):
    if os.path.exists(json_file): # 如果 JSON 已存在，刪除它以重新開始
        print(f"找到舊檔 {json_file}，將其刪除並重新開始分析。")
        os.remove(json_file)
    
    summaries = []
    graph_state = initialize_state(model, url, N, system_prompts)
    app = build_graph()
    
    # 检查并加载现有的历史分析文件
    if os.path.exists(json_file):
        print(f"找到历史文件 {json_file}，正在加载到对话记忆...")
        try:
            with open(json_file, 'r', encoding='utf-8') as f:
                existing_summaries = json.load(f)
                
            # 将JSON中的每条历史记录，作为一轮“对话”存入 memory
            for item in existing_summaries:
                user_input = f"分析時間 {item['time_window']} 商品分析結果"
                graph_state['entry_memory'].save_context({"input": user_input}, {"output": item['analysis']})
                graph_state['pv_memory'].save_context({"input": user_input}, {"output": item['pv']})
                graph_state['footprint_memory'].save_context({"input": user_input}, {"output": item['footprint']})
                graph_state['microstructure_memory'].save_context({"input": user_input}, {"output": item['microstructure']})
                graph_state['volatility_memory'].save_context({"input": user_input}, {"output": item['volatility']})
                summaries.append(item)
                
            print(f"已成功加载 {len(existing_summaries)} 条历史分析记录。")
            
        except Exception as e:
            print(f"加载历史文件时出错: {e}，将创建新的分析。")
            summaries = []
    else:
        print(f"未找到历史文件 {json_file}，将开始全新分析。")
        summaries = []

    start_date = df_all.loc[params.get('time1'): params.get('time2')]
    start_index = len(summaries)  # 例如，已分析了5个窗口，下次就从第6个开始
    total_windows = len(start_date) - k_lookback + 1
    
    # 使用 reindex 將 features 對齊到 df_all 的時間索引
    previous_day_pv = features.loc[:params.get('time1'), f"cumulative_pv_{code}"].iloc[-1]
    features = features.loc[params.get('time1'): params.get('time2')].reindex(start_date.index)
    
    for i in range(start_index, total_windows):
        df = start_date.iloc[i:i + k_lookback].copy().dropna()
        fas = features.loc[df.index].copy()
        
        target_cols = [f'open_{code}', f'high_{code}', f'low_{code}', f'close_{code}', f'pv_list_{code}', f'volume_footprint_{code}', f'volume_{code}', f'ma_diff_{code}']
        df_subset = df[target_cols].copy().dropna()
        start_time = df_subset.index[0].strftime('%Y-%m-%d %H:%M:%S')
        end_time = df_subset.index[-1].strftime('%Y-%m-%d %H:%M:%S')
        time_window = f"{start_time} to {end_time}"
        cols = [col for col in fas.columns if col.endswith(f'_{code}')]
        fa = fas[cols]

        dict_1 = {k: v for d in previous_day_pv for k, v in d.items()}
        dict_2 = {k: v for d in fa.iloc[-1][f"cumulative_pv_{code}"] for k, v in d.items()}
        s1, s2 = pd.Series(dict_1), pd.Series(dict_2)
        merged_series = s1.add(s2, fill_value=0)
        current_pv = [{k: v} for k, v in merged_series.items()]

        feature_start_time = fa.index[0].strftime('%Y-%m-%d %H:%M:%S')
        feature_end_time = fa.index[-1].strftime('%Y-%m-%d %H:%M:%S')
        feature_time_window = f"{feature_start_time} to {feature_end_time}"

        graph_state['start_time'] = start_time
        graph_state['end_time'] = end_time
        graph_state["current_pv"] = format_pv_data(current_pv, df_subset[f"close_{code}"].iloc[-1].round(2).tolist())
        graph_state["current_footprint"] = format_footprint(df_subset[f"volume_footprint_{code}"], df_subset[f"close_{code}"].round(2).tolist())
        
        # 數據處理：四捨五入 -> 轉列表 -> 轉字串 -> 去除空格 (Token 最佳化)
        open_series = f'- open: {str(df_subset[f"open_{code}"].round(2).tolist()).replace(" ", "")}\n'
        high_series = f'- high: {str(df_subset[f"high_{code}"].round(2).tolist()).replace(" ", "")}\n'
        low_series = f'- low: {str(df_subset[f"low_{code}"].round(2).tolist()).replace(" ", "")}\n'
        close_series = f'- close: {str(df_subset[f"close_{code}"].round(2).tolist()).replace(" ", "")}\n'
        volume_series = f'- volume: {str(df_subset[f"volume_{code}"].round(2).tolist()).replace(" ", "")}\n'
        ma_series = f'- ma_diff: {str(df_subset[f"ma_diff_{code}"].round(2).tolist()).replace(" ", "")}\n'
        selected_volatility = [f'vwap_{code}', f'vwap_std_{code}', f'liqudity_{code}'] 
        exclude_patterns = [f'cumulative_pv_{code}']
        selected_microstructure = [col for col in fa.columns if col not in selected_volatility]
        
        microstructure_dict = {
            col.replace(f'_{code}', ''): str(fa[col].round(2).tolist()).replace(" ", "")
            for col in selected_microstructure
            if col in fa.columns
            and not any(pattern in col for pattern in exclude_patterns)
        }
        
        product_block = close_series
        product_block += volume_series
        for feature_name, feature_values in microstructure_dict.items():
            product_block += f"- {feature_name}: {feature_values}\n"
        
        graph_state["current_microstructure"] = product_block

        volatility_dict = {
            col.replace(f'_{code}', ''): str(fa[col].round(2).tolist()).replace(" ", "")
            for col in selected_volatility
            if col in fa.columns
            and not any(pattern in col for pattern in exclude_patterns)
        }
        
        data_block = open_series
        data_block += high_series
        data_block += low_series
        data_block += close_series
        data_block += volume_series
        data_block += ma_series
        for feature_name, feature_values in volatility_dict.items():
            data_block += f"- {feature_name}: {feature_values}\n"
        
        graph_state["current_volatility"] = data_block

        print(f"[{time.strftime('%H:%M:%S')}] 第 {i+1}/{total_windows} 次调用开始")
        start_time = time.perf_counter()

        result = app.invoke(graph_state)
        
        end_time = time.perf_counter()
        print(f"[{time.strftime('%H:%M:%S')}] 第 {i+1}/{total_windows} 次调用完成，耗时: {end_time - start_time:.2f}秒\n")
        
        input = f"分析時間 {time_window} 商品分析結果"
        result['microstructure_memory'].save_context({"input": input}, {"output": result["microstructure_result"]})
        result['volatility_memory'].save_context({"input": input}, {"output": result["volatility_result"]})
        result['footprint_memory'].save_context({"input": input}, {"output": result["footprint_result"]})
        result['pv_memory'].save_context({"input": input}, {"output": result["pv_result"]})
        result['entry_memory'].save_context({"input": input}, {"output": result["analysis_result"]})

        summaries.append({
            "time": f"耗时: {end_time - start_time:.2f} 秒",
            "time_window": time_window,
            "feature_window": feature_time_window,
            "volatility": result["volatility_result"],
            "footprint": result["footprint_result"],
            "pv": result["pv_result"],
            "microstructure": result["microstructure_result"],
            "analysis": result["analysis_result"]
        })

        with open(json_file, 'w', encoding='utf-8') as f:
            json.dump(summaries, f, ensure_ascii=False, indent=4)
    
    return
    
if __name__ == "__main__":
    MODEL_NAME = "meta-llama/Llama-3.3-70B-Instruct"
    OLLAMA_BASE_URL = "http://localhost:8000/v1"
    # MODEL_NAME = "deepseek-ai/DeepSeek-R1-Distill-Qwen-32B"
    # OLLAMA_BASE_URL = "http://localhost:8001/v1"
    # MODEL_NAME = "microsoft/phi-4"
    # OLLAMA_BASE_URL = "http://localhost:8002/v1"
    # MODEL_NAME = "Valdemardi/DeepSeek-R1-Distill-Qwen-32B-AWQ"
    # OLLAMA_BASE_URL = "http://localhost:8003/v1"
    # MODEL_NAME = "stelterlab/phi-4-AWQ"
    # OLLAMA_BASE_URL = "http://localhost:8004/v1"
    # MODEL_NAME = "deepseek-ai/DeepSeek-R1-Distill-Llama-70B"
    # OLLAMA_BASE_URL = "http://localhost:8005/v1"
    params = {
        "time1": "2025-12-15 08:45:00",
        "time2": "2025-12-15 13:45:00"
    }

    system_prompts = {
        "prompt_check_volatility":textwrap.dedent("""
            你是一位專注於波動性分析的交易員，根據波動率週期切換交易策略。

            【輸入數據】
            1. 時間序列列表: 
                - 列表開頭是指標名稱, 後面資料為由左至右(由舊到新)。
                - 前面的數據僅作為背景基期, 多關注列表最後 3 筆數據（最新數據）的變化斜率。
            2. 歷史分析： 上次的預測（如：波動變大/波動變小）。

            【指標解釋】
            - K棒: 包含OHLCV。
            - vwap (成交均價): 市場平均成本。
            - ma_diff: 均線差值, 代表趨勢強度，請觀察其「數值變化趨勢」。
                - 狀態判定閾值 (Threshold): **0.15**
                - 糾結:
                    1. 「數值極小」：abs(ma_diff) 小於 0.15，代表兩線幾乎黏合。
                    2. 「走勢躺平」：近期最後 3-4 筆數據幾乎相同 (例如連續維持在 -0.10)，代表均線平行，無趨勢方向。
                    *只要滿足上述任一條件，即判定為糾結。*
                - 發散: abs(ma_diff) 的數值「持續變大」(遠離 0 軸)。代表趨勢正在增強。
                - 收斂: abs(ma_diff) 的數值「持續變小」(靠近 0 軸)，但尚未進入糾結區間。代表乖離過大後的修正。
            - vwap_std (波動標準差): *注意：我們尋找的是「二階導數」的變化，即加速度是否變慢或轉負。*
                - 數值上升：波動放大/發散。
                - 數值下降：波動收斂/壓縮。
            - liqudity (量能脈衝): 當前成交量相較於歷史的流動性水平。
                - 數值衝高後回落：代表激情過後，動能衰退。
                - 數值持續創高：代表新資金不斷湧入，動能強勁。
                
            【分析邏輯：先判斷波動，再判斷價位】
            請將輸入視為一段動態行情，並嚴格依序執行以下步驟：
             **第一步：波動性檢查**
                - **動作一：數據提取**
                    1.請務必先從輸入數據的 `ma_diff` 欄位中，提取最後三筆數值，並計算其絕對值 `abs()`。
                        - *自我檢查：確認提取的是 ma_diff 而非 vwap_std 或 liquidity。*
                    2. 掃描 `vwap_std` 和 `volume` 數據，找出 **「近期峰值 (Local Peak)」** 及其發生時間。 

                - **動作二：狀態判定 (請嚴格對號入座)**
                    請根據提取的最後三筆數值，直接匹配以下三種情境之一：
                    1. **【判定：發散 (Divergence)】**
                        - *條件*：最後三筆 abs(ma_diff) 數值呈現「由小變大」的趨勢，且最新一筆數值的 abs > 0.15。
                        - *意義*：趨勢正在增強。
                    2. **【判定：收斂 (Convergence)】**
                        - *條件*：最後三筆 abs(ma_diff) 數值呈現「由大變小」的趨勢。
                        - *關鍵例外*：只要這三筆數據中，**曾經有任何一筆** abs > 0.15，即便最新值降至 0.15 邊緣，仍必須判定為「收斂」（視為趨勢的回調修正）。
                        - *意義*：乖離過大後的修正，動能正在減弱。
                    3. **【判定：糾結 (Tangled)】**
                        - *條件*：上述兩種情況皆不符合。也就是說，最後三筆數據的 abs 全部都「持續」小於 0.15。
                        - *意義*：均線黏合，無明確趨勢。
      
                - **動作三：邏輯判定 (優先級由高到低，符合即停止)**：
                    - **狀態 S (壓縮反轉) [特殊背離]**：
                        - *條件*：`vwap_std` 正在下降或處於低檔 (壓縮)，且 MA 正在「收斂」狀態, **但是** `volume` 或 `liqudity` 卻顯著上升/創新高 (大於前三根均值)。
                        - *判讀*：此為「量大波動小」的異常狀態，代表控盤者正在鎖定籌碼進行單邊推升，或是變盤前的極致壓縮。**此時不可視為衰退，必須視為「蓄勢待發」。**

                    - **狀態 A (波動收斂/動能衰退)**：
                        - *條件*：`vwap_std` 下降 < 近期峰值，**且** `liqudity` 同步下滑或無明顯放大, MA 由「發散」轉為「收斂」(開口縮小)。
                        - *判讀*：標準差不再創新高，且量能退潮，代表劇烈變動已過，市場進入休息期。**此時定調為「收斂」。**

                    - **狀態 B (波動擴大/衝刺)**：
                        - *條件*：`vwap_std` >= 近期峰值 (創新高)，且 `當前` > `前一筆`, 且 MA 由初始的「收斂, 糾結」轉變為「發散」狀態 (開口持續擴大)。
                        - *判讀*：波動率與量能同步爬升，且均線支撐/壓力強勁，行情正在發散階段。

                    - **狀態 C (平穩/震盪)**：
                        - *條件*：無上述特徵，MA 呈現「橫向糾結」或「走平交叉」，且其餘數據平滑, 無異常量能和波瀾。
                        - *判讀*：標準的無方向盤整。
                        
            **第二步：流動性驗證**
                - 觀察 `liqudity` 的「最新值」與「前次高點」的關係：
                    - 若第一步判定為 **[狀態 S]**，則 Liquidity 必須強勢（不跌反升）。
                    - 若第一步判定為 **[狀態 A]**，則 Liquidity 應同步走弱。
                    
            **第三步：歷史預期校準**
                - 若本次判定為 **[狀態 S 壓縮蓄勢]**，則忽略歷史看空的預期，改為隨時準備順勢突破。
                - 若本次判定與歷史一致，則維持原判。
                - 若本次判定與歷史預測不符, 立刻承認錯誤, 並將本次判斷權重提高。

            **第四步：價格結構判讀 (關鍵價位與慣性分析)**
                結構分類 (請嚴格對號入座)：
                    結構 A：高檔受阻 (特徵：留長上影線，實體紅K無法收在區間高點附近, 且MA處於「偏離發散」)。
                    結構 B：低檔支撐 (特徵：留長下影線，實體黑K無法收在區間低點附近, 且MA處於「偏離發散」)。
                    結構 C：區間壓縮 (特徵：高低點區間極小，且MA處於「橫盤糾結」且沒大幅度有偏離, 價格在中間遊走)。
                    結構 D：趨勢延續 (特徵：實體長K棒，或是**連續小K棒緊貼著區間高/低點收盤**，且價格沿著發散的短均線攀升/滑落，不破均線)。

            **第五步：策略判定 (波動率 + 結構的綜合決策)**
                策略 S：靜默突襲 (Squeeze Ignition) 
                    觸發條件：[狀態 S：壓縮反轉 + MA 收斂(開口縮小)] + [結構 D：趨勢延續 (緊貼高/低點)]。
                    情境：波動率壓縮但量能放大，價格死守高/低價位不退，預示著單邊噴出。
                    操作：順勢進場 (Breakout)，嚴設起漲點為停損。
                策略 A：高空 (Reversion Short)
                    觸發條件：[狀態 A：波動收斂 + MA 收斂(開口縮小)] + [結構 A：高檔受阻]。
                    操作：偏空操作 (Short)。
                策略 B：低多 (Reversion Long)
                    觸發條件：[狀態 A：波動收斂 + MA 收斂(開口縮小)] + [結構 B：低檔支撐]。
                    操作：偏多操作 (Long)。
                策略 C：觀望 (Range Bound)
                    觸發條件：[狀態 C：平穩 + MA 橫盤糾結] + [結構 C：區間壓縮]。
                    操作：空手觀望。
                策略 D：順勢追價 (Momentum Follow)
                    觸發條件：[狀態 B：波動擴大 + MA 發散] + [結構 D：趨勢延續]。
                    操作：順勢追價。

            【輸出格式】
            請純文字輸出以下, 不要使用Json或MarkDown：
            1. 波動與結構解析:
                - 波動狀態: [波動拐點 (收斂) / 壓縮反彈 / 平穩震盪 / 趨勢爆發]
                - 均線型態: [糾結 / 發散 / 收斂]
                - 價格結構: [高檔受阻 / 低檔支撐 / 區間壓縮 / 趨勢延續]
                - 結構描述: 請嚴格依照數據執行「三段式」描述, 並依照範例句型輸出：
                    1. 波動率(vwap_std)：列出數值範圍，並明確判定狀態為「放大」、「縮小」或「持平」。
                    2. 流動性(liquidity)：列出最新值與前次高點，判定動能為「增強」、「衰退」或「無變化」。
                    3. 均線(MA_diff)：輸出「波動性檢查」所判定均線情形, 以及最近期的三筆均線數值。
                    *範例句型*：「波動率 vwap_std 為 [數值] 波動 [放大/縮小/持平]；流動性 liquidity [增強/衰退/無變化] 最新值為 [數值] 數值顯著 [放大/縮小/無變化]；當前 MA_diff [均線糾結/均線發散/均線收斂], 最後三筆為 [數值1, 數值2, 數值3]。」
            2. 交易策略焦點:
               - 策略思維: [逆勢回歸] / [區間震盪] / [順勢突破] / [靜默突襲]
               - 操作方向: [做多 / 做空 / 觀望]
               - 關鍵邏輯: 
                    *請根據當前情況，嚴格選用以下其中一種句型輸出：*
                    (情境 S：壓縮蓄勢 + MA 收斂 + 結構 D -> 靜默突襲單)
                    「偵測到 **[波動率壓縮但量能異動]** ，MA逐漸趨於收斂, 且價格結構呈現 **[緊貼高/低點]**，顯示控盤者正在鎖定籌碼進行單邊推升。此非動能衰退，而是蓄勢待發。預期價格將 **[向上/向下]** 形成單邊趨勢，建議採取 **[現價順勢建立部位]** 策略。」
                    (情境 1：波動收斂 + MA 收斂 + 高檔受阻 -> 逆勢單)
                    「由於波動率出現拐點且量能同步退潮（動能衰退），且 MA 由發散轉為 [收斂], 價格呈現 **[高檔受阻/低檔支撐]**，顯示 **[多方/空方]** 力竭。預期行情將 **[回落/反彈]** 至均值，建議採取 **[做空/做多]** 策略。」
                    (情境 2：波動爆發 + MA 發散 + 趨勢延續 -> 順勢單)
                    「波動率放大且流動性充沛，MA 呈現 [完美多頭/空頭發散], 價格結構呈現 **[趨勢延續]** 並帶量突破，建議 **[追多/追空]**。」
                    (情境 3：無波動 + MA 橫盤糾結 + 成交量少 -> 震盪單)
                    「波動低迷且無量，MA [糾結且反覆交叉]，無明確方向, 建議在區間邊緣高拋低吸或觀望。」
            3. 盤勢預測:
               - 前次預測: (簡短評論上次預測是否準確)
               - 預期行情: (描述接下來可能的路徑，例如：預期波動[收斂/擴大/平穩], MA預期[交叉/發散/收斂]。)
        """).strip(),
        "prompt_check_pv": textwrap.dedent("""
            你是一位專注於籌碼峰分析的交易員。
            
            【輸入數據說明】
            1. 當前籌碼數據: 價格與籌碼圖與當前收盤價
            2. 歷史驗證資料: 上次分析的「收盤價」與「預判劇本」(例如: 看跌至34.80)

            【分析邏輯】
            請嚴格執行以下步驟, 不要輸出思考過程, 僅輸出結論：
            
            步驟一：隱性歷史校準
                - 先比對「當前價格」是否符合「上次預判」。
                - 若預判失準（例如：預期下跌卻橫盤），請假設該處產生了新的支撐或主力護盤（隱性籌碼堆積）。
                - 此校準結果必須直接用於修正接下來的「區域狀態」判斷（例如：原判斷真空區應滑落，因失準需改判為震盪）。

            步驟二：定位與狀態識別
                - 判斷當前價格位於「籌碼密集區(HVN)」內部，還是「籌碼真空區(LVN)」或邊緣。
                - 規則 A (HVN): 若在密集區 = 容易發生「均值回歸 (盤整)」。
                - 規則 B (LVN): 若在真空區/邊緣 = 容易發生「價格滑動 (趨勢)」。

            步驟三：識別目標價位
                - 若狀態為「趨勢 (LVN)」：尋找路徑上「下一個最近的籌碼峰邊緣」作為滑動終點。
                - 若狀態為「均值回歸 (HVN)」：尋找該區域內「量能最大的價位 (POC)」作為回歸目標。

            步驟四：成交量動態修正
                - 檢查當前路徑, 相較於歷史分析結果是否有異常放量。
                - 若處於 LVN 但出現大量 = 阻力/支撐增強 -> 修正為「需修正」，狀態轉震盪。
                - 若處於 HVN 但量能萎縮且價格移出 -> 修正為「突破確立」。
            
            步驟五：策略與驗證
                - 綜合上述，決定多空方向（Long/Short/Wait）。
                - 產生一個具體的「盤勢預判」供下次驗證。
            
            【輸出格式】
            請嚴格遵守以下格式輸出（純文字）：
            1.區域狀態: [密集區/均值回歸] 或 [真空區/趨勢] (簡述理由)
            2.價格目標: [具體價位] (說明是回歸POC還是滑動至下一籌碼峰)
            3.動態修正: [無修正/需修正] (若放量導致觀點改變請說明，否則填無)
            4.預計趨勢: [空方/多方/震盪] (附帶簡易進場理由)
            5.關鍵點位: [支撐價位, 壓力價位]
            6.驗證預判: [一句話描述接下來預期發生的劇本，例如：價格將快速穿越34.80測試35.35]
        """).strip(),
        "prompt_check_microstructure": textwrap.dedent("""
            你是專精於盤口資料分析的交易員, 請根據當前輸入的所有盤口指標, 判斷是否出現「趨勢竭盡（指標背離）」或「主力吸籌/出貨」的關鍵交易訊號。
            
            【輸入數據】
            1. 盤口資料:
                - 列表開頭是指標名稱, 後面時序資料由左至右, 由舊到新, 
                - 前面的數據僅作為背景基期, 多關注列表最後 3 筆數據（最新數據）的變化。
            2. 歷史分析觀點： 前幾次對趨勢方向的預測（如：看多/看空/盤整）及其信心水準。
            
            【指標解釋】
            - close: 時序收盤價。
            - volume: 時序收盤量。
            [A. 趨勢與動能 (Trend & Momentum)]
                - vd (淨成交量差): 基礎多空力量。
                - cvd (累積成交量差): 趨勢連續性。背離判斷核心。
                - mvd (淨成交金額差): 錢的流向。大幅負值 = 資金出逃。
                - cmvd (資金動能變化率): **[加速度指標]**。
                    - 若 mvd 為負，且 cmvd 持續負值擴大：代表跌勢正在「加速」，不可接刀。
                    - 若 mvd 為負，但 cmvd 開始翻正：代表跌勢「減速」，潛在止跌契機。
                - net_sell_ratio (淨成交佔比): 衡量成交量的「純度」或「方向效率」。
                    - 高效率 (Ratio > 0.6)：代表該時段內絕大多數單子都是同一方向。若發生在「縮量」時，代表「流動性枯竭」，價格容易出現無量滑價。
                    - 低效率 (Ratio < 0.3)：代表多空互抵嚴重。若發生在「縮量」時，代表真正的「震盪」。

            [B. 籌碼結構 (Market Composition)]
                - bias_ticks (散戶意願): 
                    - 數值代表市場人氣。正值=散戶追價，負值=散戶恐慌。
                - bias_volume / bias_dollars (大戶/聰明錢意願):
                    - **[關鍵背離比對]**：必須將 `bias_ticks` 與 `bias_dollars` 對比。
                    - 情境1 (主力出貨)：bias_ticks > 0 (散戶買) 但 bias_dollars < 0 (大戶賣)。 
                    - 情境2 (主力吸籌)：bias_ticks < 0 (散戶拋) 但 bias_dollars > 0 (大戶接)。

            [C. 買賣價差失衡 (Buy & Sell Imbalance)]
                - tick_imbalance (筆數情緒): 極端值 (+/-1) 代表市場情緒過熱或恐慌。
                - volume_imbalance / dollar_imbalance (實質壓力): 
                    - 若 tick_imbalance 為負(恐慌)，但 volume_imbalance 跌幅不大，代表有人掛單吸收賣壓。
                    - 若兩者同步呈現極端負值，代表流動性被擊穿。

            [D. 交易顆粒度 (Granularity)]
                - avg_ticket_size (平均金額) / avg_volume_per_trade (平均量):
                    - **[機構確認]**：趨勢發動時，必須伴隨此數值放大。
                    - 若價格下跌，mvd 流出，且 avg_ticket_size 顯著放大：確認為**「機構砸盤」**。
                    - 若價格波動，但此數值極小：視為**「散戶噪音」**，訊號可信度低。
            
            【分析邏輯】
            請將輸入視為一段動態行情，並嚴格依序執行以下步驟與注意事項：
            **所有分析注意事項**:
            1. 最新一筆數據 (List[-1])」與「前一筆數據 (List[-2])」的變化**，而不僅是看平均值。
            2. 方向性檢測：「數值變大」不等於「看多」。必須嚴格檢查正負號以及整個時間的變化。例如：avg_ticket_size 變大且 mvd 為負，這是強烈的利空（砸盤），而非利多。

            **第一階段：多維度結構診斷**
            此階段核心目標, 透過「價格波段」與「指標波段」的位階比對，找出趨勢是否異常:
                1. 價格結構識別 (Price Structure)：
                   - 找出 `close` 列表中的最大值索引 (Index of Max Price) 與最小值索引。
                   - **位階判斷**：當前價格 (`list[-1]`) 是處於創新高(New High)、創新低(New Low)，還是處於區間震盪？
    
                2. 趨勢背離掃描 (Divergence Scan)：
                   請嚴格比對「價格極值點」與「指標對應點」的數值變化：
                   - **路徑檢查**：觀察在價格推升/下跌過程中，`cvd` (量) 與 `mvd` (資金) 是否同步墊高/降低？
                   - **高點/低點檢視**：
                     - 鎖定價格達到最高點/最低點的 Index。
                     - 檢查同一時間點的 `cvd`, `mvd`, `bias_dollars` 是否也創下該區間的極值？
                   - **異常定義 (背離判定標準)**：
                     - **[頂部背離 - 看空]**：價格創出新高 (Price New High)，但 `mvd` 或 `bias_dollars` 的峰值顯著低於前一波高點 (Indicator Lower High)。-> 意涵：價格虛漲，主力資金未跟隨。
                     - **[底部背離 - 看多]**：價格創出新低 (Price New Low)，但 `cvd` 或 `mvd` 的低點高於前一波低點 (Indicator Higher Low)，或開始走平。-> 意涵：價格下跌但賣壓枯竭。
    
                3. 主力意願與斜率分析 (Smart Money Slope)：
                   - **趨勢一致性**：對比最近 3-4 筆數據，觀察「價格斜率」與「`bias_dollars` 斜率」。
                   - **異常定義**：
                     - **主力不跟單**：價格緩步墊高，但 `bias_dollars` 呈現負值或水平線。
                     - **主力大出貨**：價格在高檔橫盤，但 `bias_volume` 突然轉為大幅負值。
    
                4. 動能健康度檢查 (Momentum Health)：
                   - 引入 `cmvd` (加速度) 進行確認。
                   - 若出現「頂部背離」，且當下 `cmvd` 由正轉負 -> 確認動能衰退，背離有效性極高。
                   - 若出現「底部背離」，且當下 `cmvd` 負值大幅縮小或翻正 -> 確認空頭減速，背離有效性極高。
    
                5. **證偽檢查 (Invalidation Check) - [最高權重]**:
                   - **目的**：確認剛才發現的背離，是否被「最新一筆數據」擊穿。
                   - **規則**：若前幾筆出現「底部背離」或「吸籌」，但最新一筆 (Last) 的 `mvd` 或 `vd` 創下整個列表的「負值新低 (New Lowest Negative)」，且 `bias_dollars` 轉負。
                   - **判定**：背離失效 (Failed Divergence)，支撐潰散，視為趨勢加速下跌訊號。

            **第二階段：觀轉折確認**
            此階段核心目標, 判斷是否出現轉折訊號:
                1. **努力與結果 (Effort vs Result)**:
                   - 觀察最新一筆的 `vd` (努力) 與 價格變化 (結果)。
                   - 若 `vd` 爆出巨量負值，且價格同步重挫 -> 努力=結果，空方強勢，無止跌跡象。
                   - 若 `vd` 爆出巨量負值，但價格跌幅極小 (Pin bar) -> 努力!=結果，被動買盤介入 (止跌)。
    
                2. **流動性與顆粒度 (Size)**:
                   - 檢查 `volume_imbalance` 是否低於 -0.2 (買盤被擊穿)。
                   - 檢查 `avg_ticket_size` 是否放大 (機構主導)。

                3. **量能效率與狀態界定**:
                    - **目的**：利用 `net_sell_ratio` 修正對縮量行情的解讀，避免誤判為強勢趨勢。
                    - **規則 A (縮量單邊 - 陰跌/盤跌)**：
                        - 若 `volume` 顯著低於平均 (近期新低)，但 `net_sell_ratio` 高 (> 0.5) 且 `mvd` 為負。
                        - **判定**：這不是強勢趨勢，而是**「流動性缺失導致的價格滑落 (Liquidity Drift)」**。
                        - **修正結論**：應定義為**「震盪 (偏空)」**，強調量能不足，後續需觀察是否補量。
                   - **規則 B (縮量混亂 - 純震盪)**：
                        - 若 `volume` 低，且 `net_sell_ratio` 低 (< 0.3)。
                        - **判定**：市場無方向，純粹噪音。
                        - **修正結論**：定義為**「區間震盪」**，建議觀望。
            
            【綜合研判與策略輸出】
            請基於上述兩個階段的發現，並結合歷史分析, 進行邏輯整合, 此階段為關鍵分析，請嚴格比對「歷史分析」與「當前診斷」：
            1. **狀態轉換檢測**：
                - **情境 A (反轉確認)**：若「歷史分析」預期走空/下跌，而「當前診斷」發現結構出現背離 + 微觀出現吸籌/竭盡訊號。
                    -> 結論：這是一個**合理的反轉訊號**，趨勢可能進行轉向, 建議平倉或反向操作。
                - **情境 B (趨勢延續)**：若「歷史分析」預期走空，而「當前診斷」顯示續跌且結構無背離 + 微觀無反轉訊號。
                    -> 結論：趨勢尚未改變，維持原有預測方向看法。
                - **情境 C (預期失敗)**：若「歷史分析」預期反彈，但「當前診斷」顯示價格無量下跌或流動性枯竭。
                    -> 結論：多頭預期失效，需重新評估。
                - **情境 D (流動性陷阱)**：若「歷史分析」有出現過大幅度的波動「當前診斷」顯示 `mvd` 大幅流出，但 `volume` 不超過近期3個時間內的平均量, 且 `net_sell_ratio` 極高。
                 -> 結論：**盤勢屬於「無量弱勢震盪」**。

            2. **一致性確認**：歷史觀點是否獲得了市場驗證？(例如：上次預期會跌，結果真的跌了 -> 觀點正確，增加當前分析的可信度)。
            
            【輸出格式】
            請純文字輸出以下，不包含 JSON 或 Markdown：
            1. **市場結構狀態**：(明確指出價格位置，並說明是否出現「價格創新高/低但指標背離」的現象。若有，請指名是 mvd 還是 bias_dollars 背離; 若成交量低迷，請標註「縮量：流動性狀態」震盪)
            2. **籌碼與流動性解讀**：(對比散戶 bias_ticks 與大戶 bias_dollars 的一致性)
            3. **微觀訊號驗證**：(檢查最新一筆是否出現「證偽/擊穿」現象，或是出現「努力與結果不符」的轉折)
            4. **交易決策建議**：
               - **趨勢方向**：(偏多 / 偏空 / 震盪 / 反轉)
               - **多單持有者**：(續抱 / 平倉)
               - **空單持有者**：(續抱 / 平倉)
               - **理由摘要**：(綜合背離結構、主力意願與最新動能狀態，給出結論)
        """),
        "prompt_check_footprint": textwrap.dedent("""
            你是一位成交量足跡圖交易員, 請根據輸入的成交量足跡圖進行分析。
            
            【輸入數據】
            1. **當前足跡數據**：按時間排序的成交量分價表。
                - **重點**：【最後3個時間點】的數據權重最高，代表當前市場的狀態。
            2. 歷史分析數據: 
                - 輸入包含前幾次的分析結論（如：「判定多頭趨勢，因為 68.5 處有空頭被套」）。
                - **用途**：請判斷是「延續」了歷史分析結果，還是出現了「反轉/失效」信號，藉此提供離場建議。
                
            【分析邏輯】
            請依照以下步驟, 總和當前和歷史分析, 尋找市場異常與交易機會：
            1. **識別異常K棒 (Key Bars)**：
               - 找出「總成交量」顯著大於前後K棒的時間點（爆量點）。
               - 觀察該K棒的收盤價位置與其最高/最低價的關係（判斷是否有長影線）。
               - 檢查「努力 vs 結果 (Effort vs Result)」： 如果出現巨量 K 棒（努力大），但隨後的 K 棒價格立刻反向吞噬或無法延伸該方向（結果差）並且出現大量成交在最高/最低價位，這不是震盪，這是強烈的反轉信號。
                
            2. **檢測「吸收 (Absorption)」與「套牢 (Trapped Traders)」**：
                - **注意**：被套牢的籌碼會轉化為強大多空方向壓力，不應視為「買賣均衡」。
                - **頂部套牢 (Bull Trap)**：如果在高價位區出現大量「買量 (Aggressive Buy)」，但價格隨後無法守住該價位並收低，代表買方被賣方被動單吸收，形成多單套牢。
                - **底部承接 (Bear Trap)**：如果在低價位區出現大量「賣量 (Aggressive Sell)」，但價格無法跌破並反彈，代表賣壓被吸收。

            3. **Delta 結構與價位分佈分析 (關鍵步驟)**：
                - **逐價位掃描**：檢查K棒內部的 Delta（買量-賣量）分佈。
                - **頂部背離 (Bearish Divergence)**：K棒創新高，但頂部的 Delta 呈現負值（主力在出貨）或是極端正值但價格收低（主力買單被吸收，多頭套牢）。
                - **底部背離 (Bullish Divergence)**：K棒創新低，但底部的 Delta 呈現正值（主力在吸籌）或是極端負值但價格收高（主力賣單被吸收，空頭套牢）。
                - **連續性**：Delta 是否呈現「單邊失衡（Imbalance）」（例如連續三個價位買量大於賣量 300% 以上，形成支撐帶）。
                - **趨勢確認**：如果在「爆量 K 棒」之後，後續 K 棒的 Delta 連續呈現相反數值，代表主控權已經轉移。
                
            4. **市場狀態判讀 (Market State)**：
                - **平衡**：成交量均勻，沒有單一方向的爆量被套牢，價格在區間內來回。
                - **失衡**：一旦出現「爆量套牢 (Trap)」，市場即脫離平衡狀態。例如：高點爆量被套 = 空頭趨勢開始（而非震盪）。
                - **量縮**： 出現套牢後，如果「量縮」，代表**「買盤枯竭」**，這意味著趨勢將延續，請勿將其解讀為市場回穩。
               
            5. **後續走勢推演**：
                - 如果發現「多頭套牢」，後續走勢判定為**偏空**，目標回測起漲點。
                - 如果發現「空頭套牢」，後續走勢判定為**偏多**，目標回測起跌點。
                - 如果無明顯套牢且 Delta 均勻，後續走勢**震盪**。
                
            6. **歷史信號驗證與離場掃描**：
                - 請調用【歷史分析紀錄】，對比當前狀態，判斷是否觸發「離場條件」:
                - **離場情境1.**
                  - *若歷史紀錄為「偏多」*：檢查當前高點是否出現「反向吸收」（賣盤巨量漲不動）或「買盤枯竭」（極低成交量創新高）。
                  - *若歷史紀錄為「偏空」*：檢查當前低點是否出現「反向吸收」（買盤巨量跌不動）或「賣盤枯竭」。
                  - **判斷**：這代表趨勢受阻，建議立即離場。
                  
                - **離場情境2.**
                  - *若歷史紀錄為「偏多」*：檢查價格是否跌回「歷史指出的支撐區/套牢區」，且伴隨買盤大量。
                  - *若歷史紀錄為「偏空」*：檢查價格是否漲回「歷史指出的壓力區」，且伴隨 賣盤大量。
                  - **判斷**：這代表原始進場邏輯（支撐/壓力）失效，建議立即離場。

            【輸出格式】
            請綜合上述分析，輸出以下結論（純文字，無 Markdown block）：
            1. 當前趨勢判定：[偏多 / 偏空 / 震盪]
            2. 關鍵籌碼現象：[描述 Delta 結構與價格關係，例如：雖然創新高，但 Delta 呈現負值，且頂部 70.5 處出現 500 口買盤吸收]
            3. 操作建議：[**重點**：結合歷史紀錄。例如：先前判斷的多頭趨勢遇到阻力，頂部出現買盤力竭，建議多單獲利平倉；或者：價格帶量跌破前期支撐 68.0，多頭失效，建議止損。]
            4. 預期後續走勢：[描述價格路徑]
            5. 關鍵觀察價位：[支撐/壓力]
        """).strip(),
        "prompt_check_entry": textwrap.dedent("""
            你是一個專業的高頻交易員, 請結合「歷史分析」與「當前數據(盤口/籌碼峰/足跡/波動率)」, 制定後續的操作策略。

            【輸入數據】
            1. **歷史分析**：先前幾次的盤勢分析結果。
            2. **波動率**：判定市場是「無波動」、「蓄勢收斂」還是「正在爆發」。
            3. **籌碼峰**：判定價格位於「真空區」還是「密集區」。
            4. **盤口**：判定是否有「主力吸籌/出貨」、「資金背離」等異常。
            5. **足跡圖**: 判定是否有出現「套牢單」。

            【分析邏輯】
            請依照以下邏輯進行判斷，不要輸出分析過程，只輸出結論：
            **步驟一：歷史校準**
                - *檢驗*：當前價格是否已觸發歷史分析的「止損位」或「劇本失效條件」？
                - *判定*：
                    - 若價格走勢與歷史預判相反（例如：看漲卻跌破關鍵支撐），**必須立即承認錯誤**，並將「當前微觀數據」的權重調至最高，視為反轉信號。
                    - 若價格符合預期，則維持原趨勢看法。
                
            **步驟二：籌碼結構重要判斷(趨勢定義)**
                - *核心規則*：比對「當前價格」與「足跡圖/籌碼峰的大量套牢區(Trapped Zone)」。
                - *判定*：
                    - 若 **價格 < 大量套牢區** (頭部有重壓)：此時任何盤口顯示的「吸籌(Absorption)」或「正向 Bias」，應解讀為 **「空頭回補」或「下跌中繼的抵抗」**，而非反轉。趨勢判定應偏向 **延續下跌**。
                    - 若 **價格 > 大量支撐區** (底部有支撐)：此時的「吸籌」才視為有效的 **多頭防守/反轉訊號**。
            
            **步驟三：波動狀態過濾**
                *區分「收斂」與「無波動」*:
                    波動收斂：指波動率降低，但均線正在收斂，通常伴隨價格受阻或背離。這屬於「變盤前兆」，嚴禁判定為「無波動」。
                    無波動：指均線糾結且數值低於閾值(小於0.15)，且無任何微觀異常。
                *攔截規則*：
                    檢查 [波動率] 中「波動與結構解析」的結果, 只要滿足以下任一條件，禁止進入 [情境 D]，必須進入步驟四進行方向判斷：
                        1. 均線差值 (MA_diff) > 閾值 (例如 0.15)。
                        1. 關鍵字：波動狀態「拐點(收斂)」、「壓縮反彈」、「高檔受阻」、「趨勢爆發」。
                    *只有當：波動狀態為「平穩震盪」, 均線為「均線糾結」且「MA 未突破閾值」時，才允許判定為 [情境 D：無波動]*。
                
            **步驟四：情境對號入座**
                請嚴格依照以下四種「共振情境」判斷當前狀態：
                *情境 A：趨勢反轉*
                    **特徵*：[波動:收斂] + [均線:收斂] + [籌碼峰位於真空區邊緣/預計滑動] + [足跡圖發現反向套牢/吸收] + [盤口出現異常買賣壓]。
                    **結論*：趨勢即將形成。**動作：進場 / 追價。**
                
                *情境 B：順勢爆發*
                    **特徵*：[波動:擴大] + [均線:發散] + [籌碼峰脫離密集區] + [盤口 Delta/CVD 同步墊高] + [無背離訊號]。
                    **結論*：趨勢強勁延續。**動作：續抱 / 拉回加碼。**
            
                *情境 C：市場震盪*
                    **特徵*：[波動:收斂] + [均線:收斂] + [籌碼峰觸及高量區(HVN)] + [微觀出現接盤/吸收 (Absorption)] + [量價背離]。
                    **結論*：動能耗盡，回測支撐。**動作：多單止盈 / 逆勢短空。**

                *情境 D：無波動*
                    **特徵*：[波動:平穩震盪] + [均線:糾結] + [MA數值低於閾值(0.15)] + [籌碼峰位於密集區/預計回歸]。
                    **結論*：無交易價值。**動作：觀望。**
            
            **步驟五：風險定價**
            請根據步驟四判定之情境，採用不同的邏輯制定止損與觀察價位：
            **1. 動能主導模式 [情境 A/B]**
                * **核心邏輯**：以「波動率」與「價格結構」為主要防守依據。忽略籌碼峰的阻力暗示，預期趨勢將穿越籌碼密集區。
                * **止損/支撐設定**：
                    * **情境 A**：以「反向套牢區（Trapped Traders）」的極限價位或突破發起點作為強支撐/止損點。
                    * **情境 B**：以「波動率擴大起點」或「脫離成本區的關鍵K線低點」作為移動止損。
                * **籌碼峰**：
                    * **僅作「關注價位」而非出場訊號**。視籌碼峰為短期換手區或中繼站，預設價格將會突破該區，不應在此過早止盈，除非出現明確的反轉訊號。
        
            **2. 價值主導模式 [情境 C/D]**
                * **核心邏輯**：嚴格遵從「籌碼峰」的邊界進行交易。
                * **止損/支撐設定**：
                    * **情境 C**：直接以高量區（HVN）的上緣與下緣作為壓力與支撐。若價格回測籌碼峰值（POC）不破，視為進場/防守依據。
                    * **情境 D**：以籌碼密集區的核心作為中軸，上下邊緣作為極限止損，區間內不做多餘動作。

            **步驟六：點位計算與邏輯校驗 (必須執行)**：
               - 根據當前結構更新點位。
               - **強制風控邏輯**：
                 - 做空：`止盈目標 < 現價/進場位 < 止損價位`
                 - 做多：`止損價位 < 現價/進場位 < 止盈目標`
               - *特別修正*：如果是趨勢延續，進場位可以是「當前市價」，但止損位應比上次更嚴謹（向獲利方向移動）。

           【輸出格式】
            請使用繁體中文輸出以下結論（純文字，無Json或Markdown）：
            1. 當前情境: [蓄勢突破 / 順勢爆發 / 市場震盪 / 無波動]
            2. 關鍵價位: [支撐:(價位) / 壓力:(價位) / 分水嶺:(價位)]
            3. 預期盤勢: [請嚴格從以下選項擇一填入]
               - 多方趨勢
               - 空方趨勢
               - 區間震盪-壓力位空
               - 區間震盪-支撐位多
               - 無波動觀望
            4. 操作行為: 
               - [方向定調]: [做多 / 做空 / 觀望]
               - [持倉者指令]: [多單續抱 / 空單續抱 / 多單平倉 / 空單平倉]
               - [無倉者指令]: [進場做多 / 進場做空 / 觀望]
            5. 預估行情價: [數值] (目標位，若是震盪盤填區間邊緣，趨勢盤填延伸目標)
            6. 止損價位: [數值] (嚴格定義，觸發即邏輯失效)
            7. 總結: [請依序描述「波動率」+「籌碼位置」+「足跡」+「盤口」的組合分析結果，給出總結與預測方向。]
               (寫法範例：「波動率[收斂/發散] + 籌碼位於[真空區/密集區] + 足跡圖顯示[異常現象]，盤口數據[異常現象]，故綜合判定預期行情：[看多/看空/震盪/無波動]」)
        """),
    }
    
    try:
        for code in CODE:
            if code == 'FZFR1':
                continue
                
            JSON_FILE = f'./history_{code}.json'
            main(df_timek1, feature, code, JSON_FILE, MODEL_NAME, OLLAMA_BASE_URL, system_prompts, k_lookback=8, N=2, **params)
            break
    finally:
        pass
```

### VP籌碼峰

```python
pv_data = df_timek1[f'pv_list_{code}']

def create_cumulative_pv_data(pv_series):
    cumulative_data = {}
    all_previous_data = {}  # 存儲所有價位的累積量
    
    for timestamp, pv_list in pv_series.items():
        # 【新增清洗步驟】
        # 如果取出的 pv_list 不是列表 (例如是 NaN 的 float)，就強制視為空列表
        if not isinstance(pv_list, list):
            pv_list = []
        
        # 複製之前累積的數據
        current_cumulative = all_previous_data.copy()
        
        # 累加當前時間點的數據
        # (因為上面已經清洗過，這裡 pv_list 保證是 list，不會報錯)
        for item in pv_list:
            for price, volume in item.items():
                price_key = float(price)  # 確保價格是數值類型
                if price_key in current_cumulative:
                    current_cumulative[price_key] += volume
                else:
                    current_cumulative[price_key] = volume
        
        # 轉換為與原始pv_list相同的格式
        cumulative_list = [{price: volume} for price, volume in current_cumulative.items()]
        cumulative_data[timestamp] = cumulative_list
        
        # 更新累積數據用於下一個時間點
        all_previous_data = current_cumulative.copy()
    
    return pd.Series(cumulative_data, index=pv_series.index)

# 使用示例
cumulative_pv_data_test = create_cumulative_pv_data(pv_data.loc[time1: time2])
close_test = df_timek1.loc[time1: time2][f'close_{code}']
```

#### 文本

```python
def format_pv_data(pv, close):
    data_list = []
    for item in pv:
        for p, v in item.items():
            data_list.append((float(p), float(v)))
    
    # 確保按價格由低到高排序
    data_list.sort(key=lambda x: x[0], reverse=True)

    # 2. 找出最大成交量 (作為繪圖的 100% 基準)
    if not data_list:
        return "無數據"
    
    max_vol = max(item[1] for item in data_list)
    max_bar_width = 40  # 設定長條圖最大長度 (字元數)

    # 3. 建立結果字串
    result_lines = []
    result_lines.append("=" * 70)
    result_lines.append(f"籌碼峰 (當前最新收盤價: {close})")
    result_lines.append(f"{'價格':<8} | {'籌碼分佈':<42} | {'量':<6}")
    result_lines.append("-" * 70)

    # 4. 逐行繪製
    for price, volume in data_list:
        # 計算長條圖長度
        bar_len = int((volume / max_vol) * max_bar_width)
        bar_str = "█" * bar_len
        
        # 標記特殊狀態
        marker = ""
        row_style = ""
        
        # 判斷是否為當前價格 (使用小數點誤差容許比較)
        is_current = abs(price - close) < 0.001
        
        # 判斷是否為籌碼峰 (例如大於最大量的 80%)
        is_peak = volume > (max_vol * 0.8)
        
        # 判斷是否為真空區 (例如小於最大量的 25%)
        is_lva = volume < (max_vol * 0.25)

        # 組合顯示字串
        vol_str = f"{int(volume)}"
        
        if is_current:
            # 當前價位：加上箭頭強調
            marker = f" <--- Close @ {price}"
            # 為了讓當前行更顯眼，可以改變 bar 的視覺 (如果環境支援) 或加強標示
            result_lines.append(f"{price:>7.2f} | {bar_str:<42} | {vol_str:<6} {marker}")
        else:
            # 根據 LVA 或 Peak 加入簡單註解 (選用)
            note = ""
            if is_peak: note = "(峰)"
            if is_lva:  note = "." # 用小點標示相對弱勢區
            
            result_lines.append(f"{price:>7.2f} | {bar_str:<42} | {vol_str:<6} {note}")

    result_lines.append("=" * 70)
    
    # 回傳單一字串以便列印
    print("\n".join(result_lines))

index = -4
format_pv_data(cumulative_pv_data_test.iloc[index], close_test.iloc[index].round(2).tolist())
```

#### 繪圖

```python
def create_animated_chip_area(pv_series):
    fig = go.Figure()
    
    # 收集所有帧的数据
    frames = []
    timestamps = list(pv_series.index)
    
    for i, timestamp in enumerate(timestamps):
        pv_list = pv_series[timestamp]
        
        # 解析数据
        prices = []
        volumes = []
        
        for item in pv_list:
            for price, volume in item.items():
                prices.append(price)
                volumes.append(volume)
        
        # 创建DataFrame并排序
        df_chip = pd.DataFrame({'price': prices, 'volume': volumes})
        df_chip = df_chip.sort_values('price')
        
        # 创建面积图帧（更符合筹码峰的概念）
        frame = go.Frame(
            data=[go.Scatter(
                x=df_chip['price'],
                y=df_chip['volume'],
                mode='lines',
                fill='tozeroy',
                line=dict(width=2, color='royalblue'),
                fillcolor='rgba(65, 105, 225, 0.4)',
                name='筹码分布',
                hovertemplate=(
                    "价格: %{x}<br>" +
                    "成交量: %{y:,.0f}<br>" +
                    "<extra></extra>"
                )
            )],
            name=str(timestamp),
            layout=dict(
                title=f"股票 {code} 筹码分布 - {timestamp.strftime('%Y-%m-%d %H:%M')}",
                xaxis=dict(range=[df_chip['price'].min() * 0.99, df_chip['price'].max() * 1.01]),
                yaxis=dict(range=[0, df_chip['volume'].max() * 1.1])
            )
        )
        frames.append(frame)
    
    # 添加初始数据
    initial_data = frames[0].data[0]
    fig.add_trace(go.Scatter(
        x=initial_data.x,
        y=initial_data.y,
        mode='lines',
        fill='tozeroy',
        line=dict(width=2, color='royalblue'),
        fillcolor='rgba(65, 105, 225, 0.4)',
        name='筹码分布',
        hovertemplate=(
            "价格: %{x}<br>" +
            "成交量: %{y:,.0f}<br>" +
            "<extra></extra>"
        )
    ))
    
    # 创建动画按钮和滑动条（与之前相同）
    play_button = dict(
        label="播放",
        method="animate",
        args=[
            None,
            {
                "frame": {"duration": 800, "redraw": True},
                "fromcurrent": True,
                "transition": {"duration": 500}
            }
        ]
    )
    
    pause_button = dict(
        label="暂停",
        method="animate",
        args=[
            [None],
            {
                "frame": {"duration": 0, "redraw": False},
                "mode": "immediate",
                "transition": {"duration": 0}
            }
        ]
    )
    
    slider_steps = []
    for i, timestamp in enumerate(timestamps):
        step = dict(
            method="animate",
            args=[
                [str(timestamp)],
                {
                    "frame": {"duration": 0, "redraw": True},
                    "mode": "immediate",
                    "transition": {"duration": 0}
                }
            ],
            label=timestamp.strftime('%m-%d %H:%M')
        )
        slider_steps.append(step)
    
    sliders = [dict(
        active=0,
        steps=slider_steps,
        currentvalue={"prefix": "时间: "},
        pad={"t": 50},
        len=0.9,
        x=0.1
    )]
    
    # 更新布局
    fig.update_layout(
        title=f"股票 {code} 筹码分布 - {timestamps[0].strftime('%Y-%m-%d %H:%M')}",
        xaxis_title="价格",
        yaxis_title="成交量",
        height=600,
        updatemenus=[dict(
            type="buttons",
            buttons=[play_button, pause_button],
            direction="left",
            pad={"r": 10, "t": 10},
            x=0.1,
            y=1.15,
            xanchor="right",
            yanchor="top"
        )],
        sliders=sliders
    )
    
    fig.frames = frames
    
    return fig

# 创建面积图版本的筹码分布
fig = create_animated_chip_area(cumulative_pv_data.loc[time1: time2])
fig.show()
```

#### 文本

```python
result_dict = {
    ts.strftime('%H:%M'): value 
    for ts, value in fp_data[f'volume_footprint_{code}'].loc[time1:time2].items()
}

def print_volume_footprint_like_pv(result_dict, close):
    # 按時間排序
    sorted_times = sorted(result_dict.keys())
    # 建立結果字串
    result_lines = []
    result_lines.append("=" * 60)
    result_lines.append("成交量足跡分析")
    result_lines.append("=" * 60)
    
    for idx, (time_key, current_close) in enumerate(zip(sorted_times, close), 1):
        price_dict = result_dict[time_key]
        
        # 標題行
        result_lines.append(f"\n[{idx}] 時間: {time_key}")
        result_lines.append("-" * 60)
        
        # 計算該時間段的統計
        total_buy = sum(v.get('buy', 0) for v in price_dict.values())
        total_sell = sum(v.get('sell', 0) for v in price_dict.values())
        total_neutral = sum(v.get('neutral', 0) for v in price_dict.values())
        total_volume = total_buy + total_sell + total_neutral
        
        # -------------------------------------------------------
        # 【修改點】在統計數據區塊中加入收盤價
        # -------------------------------------------------------
        result_lines.append(f"總項目數: {len(price_dict)} | 總成交量: {total_volume:.1f} | 收盤價: {current_close}")
        result_lines.append(f"買量: {total_buy:.1f} | 賣量: {total_sell:.1f}")
        result_lines.append("-" * 60)
        
        # 按價格降序顯示詳細數據
        sorted_prices = sorted(price_dict.items(), key=lambda x: float(x[0]), reverse=True)
        
        for i, (price, volume_data) in enumerate(sorted_prices, 1):
            buy = volume_data.get('buy', 0)
            sell = volume_data.get('sell', 0)
            
            # 決定顯示符號
            net = buy - sell
            if net > 0:
                symbol = "↑"  # 買盤較強
            elif net < 0:
                symbol = "↓"  # 賣盤較強
            else:
                symbol = "→"  # 平衡
            
            result_lines.append(f"  {i:3d}. 價格: {price:>8} {symbol} "
                              f"買:{buy:>8.1f} 賣:{sell:>8.1f}")
    
    # 將所有行合併為一個字串並回傳
    return "\n".join(result_lines)

# 使用
print(print_volume_footprint_like_pv(result_dict, df_timek1.loc[time1: time2][f'close_{code}']))
```

#### 走勢分布圖

```python
def create_footprint_scatter_layered(fp_series):
    fig = go.Figure()
    
    # 定義分層閾值
    VOLUME_LEVELS = {
        'small': (1, 50),      # 小量
        'medium': (51, 200),   # 中量
        'large': (201, float('inf'))  # 大量
    }
    
    # 準備分層數據
    buy_data = {'small': [], 'medium': [], 'large': []}
    sell_data = {'small': [], 'medium': [], 'large': []}
    neutral_data = {'small': [], 'medium': [], 'large': []}
    
    for timestamp, fp_dict in fp_series.items():
        if isinstance(fp_dict, float) and np.isnan(fp_dict):
            continue
            
        for price, data in fp_dict.items():
            buy_vol = data.get('buy', 0)
            sell_vol = data.get('sell', 0)
            neutral_vol = data.get('neutral', 0)
            
            # 分類買盤
            if buy_vol > 0:
                for level, (min_vol, max_vol) in VOLUME_LEVELS.items():
                    if min_vol <= buy_vol <= max_vol:
                        buy_data[level].append((timestamp, price, buy_vol))
                        break
            
            # 分類賣盤
            if sell_vol > 0:
                for level, (min_vol, max_vol) in VOLUME_LEVELS.items():
                    if min_vol <= sell_vol <= max_vol:
                        sell_data[level].append((timestamp, price, sell_vol))
                        break
            
            # 分類中性盤
            if neutral_vol > 0:
                for level, (min_vol, max_vol) in VOLUME_LEVELS.items():
                    if min_vol <= neutral_vol <= max_vol:
                        neutral_data[level].append((timestamp, price, neutral_vol))
                        break
    
    # 顏色和大小配置
    level_config = {
        'small': {'size': 6, 'opacity': 0.9, 'symbol': 'circle'},
        'medium': {'size': 10, 'opacity': 0.8, 'symbol': 'circle'},
        'large': {'size': 16, 'opacity': 0.7, 'symbol': 'circle'}
    }
    
    # 添加買盤分層
    for level in ['small', 'medium', 'large']:
        if buy_data[level]:
            timestamps, prices, volumes = zip(*buy_data[level])
            fig.add_trace(go.Scatter(
                x=timestamps,
                y=prices,
                mode='markers',
                name=f'買盤-{level}',
                marker=dict(
                    size=level_config[level]['size'],
                    color='green',
                    opacity=level_config[level]['opacity'],
                    symbol=level_config[level]['symbol'],
                    line=dict(width=1, color='darkgreen')
                ),
                hovertemplate=(
                    "時間: %{x}<br>" +
                    "價格: %{y}<br>" +
                    "買盤量: %{text}<br>" +
                    "級別: " + level +
                    "<extra></extra>"
                ),
                text=volumes
            ))
    
    # 添加賣盤分層
    for level in ['small', 'medium', 'large']:
        if sell_data[level]:
            timestamps, prices, volumes = zip(*sell_data[level])
            fig.add_trace(go.Scatter(
                x=timestamps,
                y=prices,
                mode='markers',
                name=f'賣盤-{level}',
                marker=dict(
                    size=level_config[level]['size'],
                    color='red',
                    opacity=level_config[level]['opacity'],
                    symbol=level_config[level]['symbol'],
                    line=dict(width=1, color='darkred')
                ),
                hovertemplate=(
                    "時間: %{x}<br>" +
                    "價格: %{y}<br>" +
                    "賣盤量: %{text}<br>" +
                    "級別: " + level +
                    "<extra></extra>"
                ),
                text=volumes
            ))
    
    # 添加中性盤分層
    for level in ['small', 'medium', 'large']:
        if neutral_data[level]:
            timestamps, prices, volumes = zip(*neutral_data[level])
            fig.add_trace(go.Scatter(
                x=timestamps,
                y=prices,
                mode='markers',
                name=f'中性盤-{level}',
                marker=dict(
                    size=level_config[level]['size'],
                    color='blue',
                    opacity=level_config[level]['opacity'],
                    symbol=level_config[level]['symbol'],
                    line=dict(width=1, color='darkblue')
                ),
                hovertemplate=(
                    "時間: %{x}<br>" +
                    "價格: %{y}<br>" +
                    "中性盤量: %{text}<br>" +
                    "級別: " + level +
                    "<extra></extra>"
                ),
                text=volumes
            ))
    
    fig.update_layout(
        title=f"股票 {code} 成交量足跡圖 (分層版)",
        xaxis_title="時間",
        yaxis_title="價格",
        height=700,
        showlegend=True,
        plot_bgcolor='white'
    )
    
    fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
    fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
    
    return fig

fig = create_footprint_scatter_layered(fp_data[f'volume_footprint_{code}'].loc[time1: time2])
fig.show()
```

#### 價量分布

```python
def create_interactive_footprint_bidirectional(fp_series, code="Unknown"):
    fig = go.Figure()
    
    # 準備步驟數據
    steps = []
    timestamps = []
    
    # 過濾掉NaN值
    for timestamp in fp_series.index:
        fp_data = fp_series[f'volume_footprint_{code}'][timestamp] # 修改：获取成交量足跡數據
        # 檢查是否為 float 且是 nan (簡單判斷)
        if isinstance(fp_data, float) and np.isnan(fp_data):
            continue
        timestamps.append(timestamp)
    
    if not timestamps:
        print("沒有有效的數據可以繪圖")
        return fig
    
    # 添加初始數據（第一個有效時間點）
    initial_timestamp = timestamps[0]
    initial_data = fp_series[f'volume_footprint_{code}'][initial_timestamp]
    initial_close = fp_series[f'close_{code}'][initial_timestamp] # 獲取收盤價
    initial_volume = fp_series[f'volume_{code}'][initial_timestamp]  # 獲取初始成交量
    
    # --- 修改點 1: 改為 False (升序：小 -> 大) ---
    initial_prices = sorted(initial_data.keys(), reverse=False)
    
    initial_buy_volumes = []
    initial_sell_volumes = []
    
    # 創建收盤價標記的數據
    close_prices_marker = []
    close_prices_values = []
    
    for price in initial_prices:
        data = initial_data[price]
        initial_buy_volumes.append(data.get('buy', 0))
        initial_sell_volumes.append(-data.get('sell', 0))
        
        # 如果是收盤價，添加標記
        if price == initial_close:
            close_prices_marker.append(price)
            # 找到對應的成交量，用於標記位置
            buy_vol = data.get('buy', 0)
            sell_vol = data.get('sell', 0)
            # 選擇較大的成交量作為標記的x位置
            close_prices_values.append(max(buy_vol, abs(sell_vol)) * 1.1) # 增加10%間距
    
    # 添加買盤
    fig.add_trace(go.Bar(
        y=initial_prices,
        x=initial_buy_volumes,
        name='買盤',
        orientation='h',
        marker_color='green',
        opacity=0.7
    ))
    
    # 添加賣盤
    fig.add_trace(go.Bar(
        y=initial_prices,
        x=initial_sell_volumes,
        name='賣盤',
        orientation='h',
        marker_color='red',
        opacity=0.7
    ))
    
    # 添加收盤價標記（黃色三角形）
    if close_prices_marker:
        fig.add_trace(go.Scatter(
            x=close_prices_values,
            y=close_prices_marker,
            mode='markers',
            name='收盤價',
            marker=dict(
                symbol='triangle-right',
                size=12,
                color='yellow',
                line=dict(color='darkorange', width=2)
            ),
            hovertemplate='收盤價: %{y}<extra></extra>'
        ))
    
    # 創建步驟
    for i, timestamp in enumerate(timestamps):
        fp_data = fp_series[f'volume_footprint_{code}'][timestamp]
        close_price = fp_series[f'close_{code}'][timestamp] # 獲取當前時間的收盤價
        volume = fp_series[f'volume_{code}'][timestamp]  # 獲取當前成交量
        
        # --- 修改點 2: 改為 False (升序：小 -> 大) ---
        prices = sorted(fp_data.keys(), reverse=False)
        
        buy_volumes = []
        sell_volumes = []
        close_marker_prices = []
        close_marker_values = []
        
        for price in prices:
            data = fp_data[price]
            buy_volumes.append(data.get('buy', 0))
            sell_volumes.append(-data.get('sell', 0))
            
            # 如果是收盤價，準備標記數據
            if price == close_price:
                close_marker_prices.append(price)
                buy_vol = data.get('buy', 0)
                sell_vol = data.get('sell', 0)
                # 選擇較大的成交量作為標記的x位置
                close_marker_values.append(max(buy_vol, abs(sell_vol)) * 1.1)
        
        step = dict(
            method="update",
            args=[
                {
                    "y": [prices, prices, close_marker_prices], # 更新三條 traces
                    "x": [buy_volumes, sell_volumes, close_marker_values]
                },
                {
                    # 修改：在標題後方添加成交量資訊
                    "title": f"股票 {code} 成交量足跡圖 - {timestamp.strftime('%Y-%m-%d %H:%M')} (成交量: {int(volume):,})"
                }
            ],
            label=timestamp.strftime('%m-%d %H:%M')
        )
        steps.append(step)
    
    # 創建滑動條
    sliders = [dict(
        active=0,
        currentvalue={"prefix": "時間: "},
        pad={"t": 50},
        steps=steps,
        len=0.9,
        x=0.1
    )]
    
    # 更新佈局 - 修改：在初始標題添加成交量
    fig.update_layout(
        title=f"股票 {code} 成交量足跡圖 - {timestamps[0].strftime('%Y-%m-%d %H:%M')} (成交量: {int(initial_volume):,})",
        yaxis_title="價格",
        xaxis_title="成交量 (左:賣盤 => 外盤吃委買, 右:買盤 => 內盤吃委賣)",
        barmode='overlay',
        height=800,
        showlegend=True,
        sliders=sliders,
        yaxis=dict(tickformat=".2f",),
        xaxis=dict(tickformat=",")
    )
    
    return fig

# 使用示例
fig = create_interactive_footprint_bidirectional(fp_data.loc[time1: time2], code)
fig.show()
```

### K

```python
def plot_interactive_multi_candlesticks(df, CODE, colors=None):
    """
    繪製多個商品的 K 線 + 成交量，垂直堆疊
    每種商品佔 2 行：(1) K線 (2) 成交量
    """
    if colors is None:
        colors = ['blue', 'red', 'green', 'purple', 'orange', 'brown']  # 可擴充

    n_codes = len(CODE)
    rows = n_codes * 2  # 每個商品：K線 + 成交量
    subplot_titles = []
    row_heights = []

    for i, code in enumerate(CODE):
        subplot_titles.extend([f"{code} K線", f"{code} 成交量"])
        row_heights.extend([0.6, 0.4])  # K線佔比大一點

    # 1. 創建子圖
    fig = make_subplots(
        rows=rows, cols=1,
        shared_xaxes=True,
        vertical_spacing=0.02,
        subplot_titles=subplot_titles,
        row_heights=row_heights
    )

    # 2. 為每個商品添加 K線 + 成交量
    for idx, code in enumerate(CODE):
        row_k = idx * 2 + 1   # K線在奇數行
        row_vol = idx * 2 + 2 # 成交量在偶數行
        color = colors[idx % len(colors)]

        # K線
        fig.add_trace(go.Candlestick(
            x=df.index,
            open=df[f'open_{code}'],
            high=df[f'high_{code}'],
            low=df[f'low_{code}'],
            close=df[f'close_{code}'],
            name=f"{code} K線",
            increasing_line_color='red', decreasing_line_color='green'
        ), row=row_k, col=1)

        # 成交量
        fig.add_trace(go.Bar(
            x=df.index,
            y=df[f'volume_{code}'],
            name=f"{code} 成交量",
            marker_color=color,
            opacity=0.6
        ), row=row_vol, col=1)

    # 3. 更新佈局
    fig.update_layout(
        title_text='多商品 K 線 + 成交量 對比圖',
        height=400 + 300 * n_codes,  # 動態高度
        showlegend=False,
        # 所有 x 軸設為 category，避免補空白日期
    )

    # 動態設定所有 xaxis 為 category
    for i in range(1, rows + 1):
        fig.update_xaxes(type="category", row=i, col=1)
        fig.update_xaxes(rangeslider_visible=False, row=i, col=1)

    # 設定 Y 軸標題
    for idx, code in enumerate(CODE):
        row_k = idx * 2 + 1
        row_vol = idx * 2 + 2
        fig.update_yaxes(title_text=f"{code} 價格", row=row_k, col=1)
        fig.update_yaxes(title_text=f"{code} 成交量", row=row_vol, col=1)

    # 顯示
    fig.show()
    
plot_interactive_multi_candlesticks(df_timek1.loc[time1: time2], CODE)
```
